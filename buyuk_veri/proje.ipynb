# Veri Ön İşleme

```python
# spark_df içerisinde bulunan eksik gözlemlerden direkt kurtuluyoruz
spark_df = spark_df.dropna()
```

```python
# yeni bir değişken ekleyelim
spark_df = spark_df.withColumn("age_kare",spark_df.age**2)
```

```python
spark_df.show(3)
```

bağımlı ve bağımsız değişkenleri belirli bir standartta bir araya getirip oluşturup, makine öğrenmesi fonksiyonlarına göndeririz 

```python
# bağımlı değişken string değil ama string'miş gibi çekeceğiz. string muamelesi yapacağız.
# çünkü gerçek hayatta böyle gelir.
# bu veri setinde nümerik ve modellemeye hazır bir formatta
```

```python
from pyspark.ml.feature import StringIndexer
# string dönüştürücü
# bağımlı değişkenlere label
# bağımsız değişkenlere feature
# inputCol: okuyacağı değişkenin ismi
stringIndexer = StringIndexer(inputCol = "churn", outputCol="label")
```

Bu kod, PySpark'ın MLlib kütüphanesini kullanarak bir veri çerçevesindeki (DataFrame) bir sütunu sayısal bir etiket sütununa dönüştürmek için kullanılan `StringIndexer` sınıfını kullanır. Bu kodun ayrıntılı açıklaması şu şekildedir:

1. `StringIndexer` sınıfı, bir sütundaki kategorik (string) değerleri sayısal indekslere dönüştüren bir işleve sahiptir. Özellikle sınıflandırma modelleri gibi bazı makine öğrenimi algoritmaları, girdi verilerini sayısal formatta gerektirir.

2. `inputCol` parametresi, dönüştürmek istediğiniz kategorik sütunun adını belirtir. Bu durumda, "churn" adlı sütunun sayısal bir etikete dönüştürülmesi isteniyor.

3. `outputCol` parametresi, dönüştürülmüş verinin saklanacağı yeni sütunun adını belirtir. Bu durumda, dönüştürülen etiketler "label" adlı yeni bir sütunda saklanacaktır.

Bu kod, özellikle sınıflandırma problemleri için kullanışlıdır. Örneğin, bir müşterinin hizmeti iptal etme olasılığını tahmin eden bir model geliştirmek istediğinizi düşünün. "churn" sütunu müşterilerin hizmeti iptal edip etmediğini belirtiyor olabilir (örneğin, "Evet" veya "Hayır"). Ancak, bu sütunun doğrudan kullanılabilmesi için sayısal bir etikete ihtiyaç vardır, bu nedenle `StringIndexer` kullanılarak bu dönüşüm gerçekleştirilir. Bu dönüşüm, veri çerçevesini makine öğrenimi modellerine uygun hale getirmenize yardımcı olur.

girmiş olduğumuz değişkeni yakalayıp(churn) bunu label olarak bir dönüştürecek
(bu veri setindeki churn değişkeni 0 ve 1 dir. nümerik ve modellemeye hazır bir formatta)

```python
?stringIndexer
```

* A label indexer that maps a string column of labels to an ML column of label indices.
* string değişkenlerin etiketlerini bir ml modeli için label indekslerine çevirmesini ifade ediyor
* If the input column is numeric, we cast it to string and index the string values.
* eğer input değişkeni nümerik ise bu string'e çevrilir ve string ifadeler indekslenir

* The indices are in [0, numLabels). By default, this is ordered by label frequencies
so the most frequent label gets index 0.
* 0 ve 1 sınıfının sınıflandırma problemlerinde önemli bir problemdir. burada ön tanımlı olarak sınıf etiketlerinin en çok gözlenen değerinin 0 olarak etiketlendiği bilgisi verilmiştir.

```python
indexed = stringIndexer.fit(spark_df).transform(spark_df)
```

```python
indexed.dtypes
```

```python
indexed.show(3)
```

```python
# label -> int yapacağım
spark_df = indexed.withColumn("label", indexed["label"].cast("integer"))
```

```python
spark_df.dtypes
```

```python
# bağımsız değişkenlerin ayarlanması
```

```python
from pyspark.ml.feature import VectorAssembler
```

```python
spark_df.columns
```

## Bağımsız Değişkenler

```python
bg = ["age","total_purchase","account_manager","years","num_sites"]
bg
```

```python
# VectorAssembler ile bağımsız değişkenlerden bir vektör oluşturacağız.
vectorAssembler = VectorAssembler(inputCols=bg, outputCol="features") # vektör dönüştürme nesnesi
```

```python
va_df = vectorAssembler.transform(spark_df)
```

## Final Modeli

```python
# label ile bağımlı değişkeni alır
# feature ile bağımsız değişkenleri alır
final_df=va_df.select(["features","label"])
```

```python
final_df.show()
```

pyspark ml öğrenmesi fonksiyonlara gönderilmesi gereken sabit bir yapı var bu feature ve label yapısı

## test ve train

```python
splits = final_df.randomSplit([0.7,0.3])
```

```python
train_df=splits[0]
train_df
```

```python
test_df=splits[1]
test_df
```

# GBM ile Müşteri Terk Modellemesi

```python
from pyspark.ml.classification import GBTClassifier
```

```python
# model nesnesini oluşturdum
gbm = GBTClassifier(maxIter=10, featuresCol="features", labelCol="label")
```

```python
# modeli kurulumu
gbm_model = gbm.fit(train_df)
```

```python
s = gbm_model.transform(test_df)
s
```

* probability: olasılık
* features: bağımsız değişkenler
* label: bağımlı değişkenler(y gerçek değerler)
* prediction: y tahmin edilen değerler

```python
ac = s.select("label","prediction")
```

```python
# ilkel test hatası validate edilmemiş
# accuracy
ac.filter(ac.label == ac.prediction).count() / ac.count()
```

```python
# validate edilmesi (doğrulanması) cross validation yönteminin kullanılması
```

## Model Tuning

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
```

```python
# değerlendirme yöntemimiz: BinaryClassificationEvaluator
evaluator = BinaryClassificationEvaluator()
```

```python
paramGrid = (ParamGridBuilder()
            .addGrid(gbm.maxDepth,[2,4,6]) # max derinlik
            .addGrid(gbm.maxBins, [20,30])
            .addGrid(gbm.maxIter, [10,20])
            .build())
```

```python
# estimator: model nesnesi
# numFolds: 10 katlı olacak
cv = CrossValidator(estimator=gbm, estimatorParamMaps=paramGrid, evaluator=evaluator, numFolds=10) 
```

```python
# bu parametre setlerine karşılık optimum oluşabilecek model burada kurulmuş olacak
```

```python
# uygulayalım
cv_model = cv.fit(train_df)
```

```python
s = cv_model.transform(test_df)
s
```

```python
ac = s.select("label","prediction")
```

```python
# accuracy
ac.filter(ac.label == ac.prediction).count() / ac.count()
```

%88 çıktı çok başarılı tahmin yapıldı. 

```python
# eğri altında kalan alana bakalım
evaluator.evaluate(s)
```
