# Histogram ve Yoğunluk

- histogram ve yoğunluk grafikleri sayısal değişkenler için kullanılan sayısal değişkenlerin dağılımını ifade etmek için kullanılan veri görselleştirme teknikleridir.

- histogram, sayısal değişkenin/sürekli değişkenin dağılımıyla ilgili bilgi verir.

```python
import seaborn as sns
```

```python
diamonds=sns.load_dataset("diamonds")
```

```python
df=diamonds.copy()
df.head()
```

- histogram sayısal değişkenin değerlerini belirli aralıklara böler ve belirli aralıklardaki ilgili değerlerin gözlemle frekanslarını yansıtır. 
- örneğin 0 ile 100 arasında bu kadar değer gözlenmiş
- değişkenin belirli değerlerdeki dağılımıyla ilgili bilgi vermiş olur.

```python
# displot, dağılım görselleştirmek için kullanılır.
# kde: yoğunluk fonksiyonu grafiğin üzerine koyulsun mu?
sns.distplot(df.price, kde=False);
```

- grafiğe baktığımızda price değişkeninin yayıldığını dolayısıyla std yüksek olduğunu görebiliriz.
- çarpık olduğunu görebiliriz.
- medyanı ile ortalamasının arasında fark olduğunu görebiliriz.

```python
# price'in betimsel istatistiklerine bakalım
df["price"].describe()
```

- çeyrek1(25%): 950.000000 İlk çeyrek veya alt çeyrek olarak da bilinir.
- medyan(50%): 2401.000000 ortanca 
- çeyrek3(75%): 5324.250000 Üçüncü çeyrek veya üst çeyrek

- "price" (fiyat) değişkeninin sola çarpık (negatif çarpık) olduğunu söyleyebiliriz. Sola çarpıklık, veri dağılımının sağ tarafta uzun kuyruklu ve sol tarafta yoğun olduğu durumu ifade eder. Bu, verilerin düşük değerlerde yoğunlaştığı ve yüksek değerlerde daha seyrek olduğu anlamına gelir.

- Bu sonuca ulaşmamızın nedeni, istatistikler arasında ortalama (mean) ve medyan (50. yüzdelik) değerlerine bakarak anlaşılabilir. Eğer ortalama (mean) medyandan (median) daha küçükse, veri dağılımı sola çarpıktır. Eğer ortalama (mean) medyandan (median) daha büyükse, veri dağılımı sağa çarpıktır.

- Örneğin, burada ortalama "3932.799722" ve medyan "2401.000000" olarak verilmiş. Dolayısıyla, ortalama medyandan daha büyük olduğu için "price" (fiyat) değişkeni sola çarpık (negatif çarpık) olarak değerlendirilebilir. Bu, daha düşük fiyatlı verilerin daha yoğun olduğunu ve yüksek fiyatlı verilerin daha nadir olduğunu gösterir. Sola çarpıklığın nedenleri arasında özel durumlar, sınırlamalar veya dışlayıcı değerler (outliers) olabilir.

```python
sns.histplot(df.price,kde=False);
```

```python
# bins: artırırsak hassasiyet artar, x eksenindeki değerlerin temsil edilişi düşer.
sns.distplot(df.price,bins=1000,kde=False);
```

```python
sns.histplot(df.price,bins=1000,kde=False);
```

```python
# bins: 10 yaptığımızda okunması daha kolay bir hale geldi
sns.distplot(df.price, bins=10, kde=False);
```

- bins: aralıklarla ilgilidir.
- bins, 10 ise 10 tane çubuk olacak şekilde bu grafik oluşturulur.
- bins, 100 ise 100 tane çubuk olacak şekilde bu grafik oluşturulur.
- bins'i küçülttüğümüzde çubukların genişliğinin arttığını gözlemleriz. dolayısıyla barındırdıkları değerlerin artmasını bekleriz.
- bu da örneğin price 0 ile 2400 arasında 25000 tane değer vardır.

```python
sns.histplot(df.price,bins=10,kde=False);
```

## Histogram'ın üzerine yoğunluk grafiği yapalım

```python
# sns.distplot(df.price,kde=True);
sns.distplot(df.price);
```

- görülen bu çizgi olasılık yoğunluk fonksiyonunun grafiğidir.
- y ekseninin ölçeği değişmiş
- histogram ile yoğunluğun birlikte gösterilmesi istendiğinde y ekseninde ölçekler birbirinden farklı olacağından dolayı görselleştirme imkanı olmayacağından seaborn olasılık yoğunluk fonksiyonunun ölçeğine göre 0 ile 1 arasına indirgeyip ikisini birden gösterir.

```python
sns.histplot(df.price, kde=True);
```

### Sadece yoğunluk grafiği için

```python
sns.distplot(df.price,hist=False);
```

```python
sns.displot(df.price, kde=True);
```

```python
# Ayrı olarak kde grafiğini görmek için 
sns.displot(df.price, kind = "kde");
```
