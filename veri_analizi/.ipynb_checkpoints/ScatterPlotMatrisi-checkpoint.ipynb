# Scatterrplot Matrisi

- Scatterplot, değişkenler arasındaki ilişkiyi ifade eden bir saçılım grafiğidir.
- dolayısıyla veri seti içerisindeki tüm sayısal değişkenler için matris formunda bu ilişkiyi ifade ettiğimizde bir scatterplot matrisi oluşturmuş oluyoruz.

```python
import seaborn as sns
```

```python
iris=sns.load_dataset("iris")
```

```python
df=iris.copy()
df.head()
```

- bu veri seti içerisinde bazı çiçek türleri var ve bu türlerin özelliklerini ifade eden değişkenler var.
- yapraklarının uzunluğu, genişliği vb.
- species, türleri ifade eder. bu değişkene odaklanacağız. nominal

```python
df.dtypes
```

```python
df.shape # 150 gözlem, 5 değişken
```

```python
sns.pairplot(df);
```

- değişkenlerin birbirleri arasındaki ilişkiler görselleştirilmiştir.
- sütun grafikleri, veri setinde yer alan değişkenlerin dağılımlarını ifade ediyor.
- diğer grafikler ile değişkenler arasındaki ilişki ifade edilmiştir.
- scatterplot görsellerinde, saçılım grafiği görsellerinde eğer oluşturulmuş olan grafik bir toz bulutu şeklindeyse veya yapısal bir formu yoksa bu o iki değişken arasında bir ilişki olmadığı anlamına gelir.
- gözlem noktaları incelendiğinde bunlar birbirinden farklı köşelerde kümeleniyorsa bu durumda çaprazlama durumundan kaynaklı olan, çaprazlama durumuyla gözlemleyebilecek olduğumuz oluşan bu yapıları, farklı farklı kümelenmeleri ifade eden bazı alt gruplar vardır. bunların göz önünde bulundurulması gerekir.

```python
sns.pairplot(df, hue="species"); # tür
```

- eğer çeşitli noktalarda kümelenmeler varsa bu bazı farklı yapılanmaları ifade eder.

```python
sns.pairplot(df, hue="species", markers=["o","s","D"]);
```

- markers işaretleriyle türlerin, sınıfların şeklini değiştirdik.

```python
# grafiğe doğru ekleyelim
sns.pairplot(df, kind="reg");
```

```python
# türlere göre ayıralım
sns.pairplot(df, kind="reg", hue="species");
```

- (2,0) daki grafikte petal_length ile sepal_length arasındaki ilişkide 
- versicolor ve virginica türleri için çok şiddetli, pozitif bir korelasyon, ilişki var.
- fakat setosa'ya türü göz önünde bulundurulduğunda bu ilişkinin olmadığı gözlenir. eğim yok.
