# Doğrusal İlişkinin Gösterilmesi

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
```

```python
df=tips.copy()
df.head()
```

```python
sns.lmplot(x="total_bill", y="tip", data=df);
```

- iki değişken arasındaki doğrusal ilişki
- doğru etrafındaki açık renk mavi sapmaları ifade eder.
- total_bill'in 17 civarına denk gelen kısımlarda açık mavi yok ve mavi yuvarlaklar yani değerler birbirine çok yakındır.
- arttıkça değerlerin uzaklaştığı görüldüğünden standart sapma artar ve açık mavi büyür.
- hesap arttıkça, tip artmıştır.

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=df);
```

- sigara içenlerin artışının, içmeyenlere göre fazla olmadığı görülür. Eğimin azaldığı görülür.

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", data=df);
```

- col argümanı verilmiş olan bir kategorik değişkene göre iki farklı grafik oluşturmuş oldu.
- öğünlere göre eğimlerin nasıl değiştiğini gözlemliyoruz.

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", row="sex", data=df);
```
