j colorun fiyatının fazla cıkmasının sebebi

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("diamonds")
df.head()
```

```python
# sadece J color olanlardan bir data set
j_data=df[df["color"]=="J"]
```

```python
# sadece D colorlardan oluşan bir dataset
d_data=df[df["color"]=="D"]
```

```python
# D colorda kaç farklı karat değeri var
d_data["carat"].nunique()
```

```python
# J colorda kaç farklı karat değeri var
j_data["carat"].nunique()
```

```python
# J colorda kaç farklı karat değeri var? bunlardan oluşan bir liste yapalım.
j_list=list(j_data.carat.unique())
j_list
```

```python
# D colorda kaç farklı karat değeri var? bunlardan oluşan bir liste yapalım.
d_list=list(d_data.carat.unique())
d_list
```

```python
# j color da her karat değeri için kaç adet pırlanta içeriyor
j_adet=[]

for i in j_list:
    x=j_data[j_data["carat"]==i]
    adet=len(x)
    j_adet.append(adet)
    
j_adet
```

```python
# d color da her karat değeri için kaç adet pırlanta içeriyor
d_adet=[]

for i in d_list:
    x=d_data[d_data["carat"]==i]
    adet=len(x)
    d_adet.append(adet)
    
d_adet
```

```python
## bunlardan yeni birer dataframe oluşturalım
j=pd.DataFrame({"Karat_Değeri":j_list,"Adet":j_adet})
j
```

```python
## bunlardan yeni birer dataframe oluşturalım
d=pd.DataFrame({"Karat_Değeri":d_list,"Adet":d_adet})
d
```

### daha iyi gözlemleyebilmek için yeniden indexleyelim

```python
# d color için indexleme değeri
index_d=(d["Karat_Değeri"].sort_values(ascending=False)).index.values
d=d.reindex(index_d)
d.head()
```

```python
# j color için indexleme değeri
index_j=(j["Karat_Değeri"].sort_values(ascending=False)).index.values
j=j.reindex(index_j)
j.head()
```

yani birinde en büyük karat 5.01 diğerindeki de 3.4 renk olarak daha iyi olsada karat büyüdüğü zaman pahalı olabiliyor.

- nunique() = bir dataframin sutunlarındaki eşşiz değerlerin sayısını verir.
- sort_values () = seçtiğiniz sutundaki değerleri sıralar
- index = dataframe de şeçtiğiniz satır veya satırların indexlerini alır sonuna values ekleyincede hocam o indekslerin sayısal değerlerini alıyor.
