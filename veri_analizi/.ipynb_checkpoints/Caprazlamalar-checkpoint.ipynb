```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
```

```python
diamonds = sns.load_dataset("diamonds")
```

```python
df=diamonds.copy()
```

```python
cut_kategoriler=["Fair","Good","Very Good","Premium","Ideal"]
df.cut=df.cut.astype(CategoricalDtype(categories=cut_kategoriler, ordered=True))
```

```python
# color: renk (from J (worst) to D (best))
color_kategoriler=["J","I","H","G","F","E","D"]
df.color=df.color.astype(CategoricalDtype(categories=color_kategoriler, ordered=True))
```

```python
df.head()
```

```python
# catplot: kategorik değişkenleri görselleştirmek için kullanılan bir fonksiyondur.
sns.catplot(x="cut",y="price",data=df);
```

- Fair: price 8000 ve aşağısında noktaların yoğunluğu arttığı görülür.
- cut(kalite) ve price arttıkça yoğunluğun arttığı görülür.

```python
sns.barplot(x="cut",y="price",hue="color",data=df);
```

- price 0-18000 arasinda olmasina ragmen yukaridaki tabloda 0-6000 arasinda gosterilmesinin sebebi
burada gorulen price eksenindeki degerler iki kategorik degisken birarada bulundugunda 2 groupby islemi
sonrasinda price degiskeninin ortalamasi
- uzerine koyulan cubuklar ise standart sapma degerleri
standrat hata olarak da degerlendirilebilir yani ortalama etrafindaki dagilimi gosteren bir olcut

simdi olanlari dogrulayalim:
- eksen bozuldu 0-6000 arasina dondu: bunun bir toplulastirma islemi oldugu bilinir.
- dogrulama

df.groupby(["cut","color"])["price"].mean()

Bu kod satırı, "df" DataFrame'indeki "cut" ve "color" sütunlarına göre gruplandırma yaparak "price" sütununun ortalamasını hesaplıyor.

Aşağıda bu kodun nasıl çalıştığını adım adım açıklayayım:

1. `df`: Seaborn kütüphanesi ile yüklediğimiz "diamonds" veri kümesinin bir kopyasını içeren DataFrame.

2. `groupby(["cut", "color"])`: Bu kısım, "cut" ve "color" sütunlarına göre verileri gruplamak için kullanılan işlevi çağırıyor. Yani, her bir benzersiz "cut" ve "color" değeri için aynı kategoriye sahip satırlar bir araya getirilir ve aynı gruptaki satırlar birbirleriyle ilişkilendirilir.

3. `["price"]`: Bu kısım, gruplanan veriler içinden sadece "price" sütununu seçmemizi sağlar. Yani, sadece "price" sütunundaki değerlerle ilgileniriz.

4. `.mean()`: Son olarak, "price" sütunundaki değerlerin ortalamasını hesaplamak için `mean()` işlevini çağırıyoruz. Bu, her bir "cut" ve "color" grubundaki elmasların fiyatlarının ortalamasını verir.

Sonuç olarak, bu kod, her bir farklı "cut" ve "color" değeri kombinasyonu için elmasların ortalama fiyatını hesaplar. Bu tür analizler, veri kümesi içindeki farklı kategoriler arasındaki fiyat farklarını anlamak için oldukça faydalıdır.

```python
df.groupby(["cut","color"])["price"].mean()
```

```python
# sadece D color pırlanta içeren bir dataframe oluşturalım
df[df["color"]=="D"]
```

```python
# Sonrasında Bu dataframe i kesim sınıflarına göre gruplayalım
df[df["color"]=="D"].groupby("cut")
```

```python
df[df["color"]=="D"].groupby("cut").mean()
```

```python
df[df["color"]=="D"].groupby("cut").describe().T
```

```python
# sadece carat sütununa ulaşmak
((df[df["color"]=="D"].groupby("cut")["carat"]).describe())
```

```python
# Bu değerlerden de standart sapma değerine ulaşmak için ise 
# sonuna köşeli parantez içine "std" yazmanız yeterli olacaktır
((df[df["color"]=="D"].groupby("cut")["carat"]).describe())["std"]
```

```python
# Kısa Yolu
df[df["color"]=="D"].groupby("cut").std()
```

```python
# datasette ki her kaç tane renk varsa hepsine erişmek için bir liste oluşturalım. 
# Eşsiz örnekleri pandasta unique() fonksiyonu ile bulabilmekteyiz
liste= df["color"].unique()
liste
```

```python
# her bir renkte kaç eleman var onu bulalım
for i in liste:
    print(i,"renginde",len(df[df["color"]==i]))
```

```python
standart_sapma=df[df["color"]=="D"].groupby("cut")["carat"].std().tolist()
standart_sapma
```

```python
kesimler=df["cut"].unique()
kesimler
```

```python
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=kesimler,y=standart_sapma)
plt.title("carat için Rengi D olan Pırlantaların Kesimlerine Göre Aralarındaki Standart Sapma Farkları")
plt.show()
```

```python
# tüm renkler için
for i in liste:
    sns.barplot(x=df["cut"].unique(),
                y=df[df["color"]==i].groupby("cut")["carat"].std().tolist());
    plt.title("carat için Rengi {} olan Pırlantaların Kesimlerine Göre Aralarındaki Standart Sapma Farkları".format(i))
    plt.show()
```
