```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
```

```python
diamonds = sns.load_dataset("diamonds")
```

```python
df=diamonds.copy()
```

```python
cut_kategoriler=["Fair","Good","Very Good","Premium","Ideal"]
df.cut=df.cut.astype(CategoricalDtype(categories=cut_kategoriler, ordered=True))
```

```python
# color: renk (from J (worst) to D (best))
color_kategoriler=["J","I","H","G","F","E","D"]
df.color=df.color.astype(CategoricalDtype(categories=color_kategoriler, ordered=True))
```

```python
# catplot: kategorik deÄŸiÅŸkenleri gÃ¶rselleÅŸtirmek iÃ§in kullanÄ±lan bir fonksiyondur.
sns.catplot(x="cut",y="price",data=df);
```

df.groupby(["cut","color"])["price"].mean()

Bu kod satÄ±rÄ±, "df" DataFrame'indeki "cut" ve "color" sÃ¼tunlarÄ±na gÃ¶re gruplandÄ±rma yaparak "price" sÃ¼tununun ortalamasÄ±nÄ± hesaplÄ±yor.

AÅŸaÄŸÄ±da bu kodun nasÄ±l Ã§alÄ±ÅŸtÄ±ÄŸÄ±nÄ± adÄ±m adÄ±m aÃ§Ä±klayayÄ±m:

1. `df`: Seaborn kÃ¼tÃ¼phanesi ile yÃ¼klediÄŸimiz "diamonds" veri kÃ¼mesinin bir kopyasÄ±nÄ± iÃ§eren DataFrame.

2. `groupby(["cut", "color"])`: Bu kÄ±sÄ±m, "cut" ve "color" sÃ¼tunlarÄ±na gÃ¶re verileri gruplamak iÃ§in kullanÄ±lan iÅŸlevi Ã§aÄŸÄ±rÄ±yor. Yani, her bir benzersiz "cut" ve "color" deÄŸeri iÃ§in aynÄ± kategoriye sahip satÄ±rlar bir araya getirilir ve aynÄ± gruptaki satÄ±rlar birbirleriyle iliÅŸkilendirilir.

3. `["price"]`: Bu kÄ±sÄ±m, gruplanan veriler iÃ§inden sadece "price" sÃ¼tununu seÃ§memizi saÄŸlar. Yani, sadece "price" sÃ¼tunundaki deÄŸerlerle ilgileniriz.

4. `.mean()`: Son olarak, "price" sÃ¼tunundaki deÄŸerlerin ortalamasÄ±nÄ± hesaplamak iÃ§in `mean()` iÅŸlevini Ã§aÄŸÄ±rÄ±yoruz. Bu, her bir "cut" ve "color" grubundaki elmaslarÄ±n fiyatlarÄ±nÄ±n ortalamasÄ±nÄ± verir.

SonuÃ§ olarak, bu kod, her bir farklÄ± "cut" ve "color" deÄŸeri kombinasyonu iÃ§in elmaslarÄ±n ortalama fiyatÄ±nÄ± hesaplar. Bu tÃ¼r analizler, veri kÃ¼mesi iÃ§indeki farklÄ± kategoriler arasÄ±ndaki fiyat farklarÄ±nÄ± anlamak iÃ§in oldukÃ§a faydalÄ±dÄ±r.

```python
df.groupby(["cut","color"])["price"].mean()
```

```python
# sadece D color pÄ±rlanta iÃ§eren bir dataframe oluÅŸturalÄ±m
df[df["color"]=="D"]
```

```python
# SonrasÄ±nda Bu dataframe i kesim sÄ±nÄ±flarÄ±na gÃ¶re gruplayalÄ±m
df[df["color"]=="D"].groupby("cut")
```

```python
df[df["color"]=="D"].groupby("cut").describe().T
```

```python
# sadece carat sÃ¼tununa ulaÅŸmak
((df[df["color"]=="D"].groupby("cut")["carat"]).describe())
```

```python
# Bu deÄŸerlerden de standart sapma deÄŸerine ulaÅŸmak iÃ§in ise 
# sonuna kÃ¶ÅŸeli parantez iÃ§ine "std" yazmanÄ±z yeterli olacaktÄ±r
((df[df["color"]=="D"].groupby("cut")["carat"]).describe())["std"]
```

```python
# KÄ±sa Yolu
df[df["color"]=="D"].groupby("cut").std()
```

```python
# datasette ki her kaÃ§ tane renk varsa hepsine eriÅŸmek iÃ§in bir liste oluÅŸturalÄ±m. 
# EÅŸsiz Ã¶rnekleri pandasta unique() fonksiyonu ile bulabilmekteyiz
liste= df["color"].unique()
liste
```

```python
# her bir renkte kaÃ§ eleman var onu bulalÄ±m
for i in liste:
    print(i,"renginde",len(df[df["color"]==i]))
```

```python
standart_sapma=df[df["color"]=="D"].groupby("cut")["carat"].std().tolist()
standart_sapma
```

```python
kesimler=df["cut"].unique()
kesimler
```

```python
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=kesimler,y=standart_sapma)
plt.title("carat iÃ§in Rengi D olan PÄ±rlantalarÄ±n Kesimlerine GÃ¶re AralarÄ±ndaki Standart Sapma FarklarÄ±")
plt.show()
```

```python
# tÃ¼m renkler iÃ§in
for i in liste:
    sns.barplot(x=df["cut"].unique(),
                y=df[df["color"]==i].groupby("cut")["carat"].std().tolist());
    plt.title("carat iÃ§in Rengi {} olan PÄ±rlantalarÄ±n Kesimlerine GÃ¶re AralarÄ±ndaki Standart Sapma FarklarÄ±".format(i))
    plt.show()
```
