# Binom Dağılımı

Binom dağılımı, bağımsız n deneme sonucu k başarılı olma olasılığı ile ilgilenildiğinde kullanılan dağılımdır.

- f, olasılık hesaplamak için kullanılan bir fonksiyondur.
- E, dağılımın beklenen değeri
- Var, dağılımın varyansı
- p, başarılı olma olasılığı
- n, deneme sayısı

Bernoulli'de başarılı-başarısız olmak üzere 2 sonuç vardı. burada yine başarılı-başarısız olmak üzere 2 sonuç var. farkı ise n deneme vardır. n defa bir şey deneniyor olacak ve bu deneme sonucunda başarılı olmak olayı ile ilgileniyor olacağız.
Bu bernoulli'nin n defa gerçekleştirilmiş halidir. Deneyler bağımsız olmalı ve aynı koşullar altında gerçekleştirilmelidir.

p, başarı olasılığı ; 1-p ise başarısızlık olasılığı bu gerçekleştirilecek olan her bağımsız deneme için aynıdır, eşittir.

Bir kesikli rassal değişkenin binom dağılımı gösterilmiştir.

Rassal değişkenler bazı değerleri aldığında hangi olasılıkları üreteceği bilgisi olasılık dağılımı formuyla sunulmuştur.

x rassal değişkeni binom dağılımına sahip, 
n=10, p=0.3

binom dağılımına sahip x rassal değişkeni 1 aldığında, olasılık değeri y eksenine bakıldığında 0.10 ile 0.15 arasında olduğu görülür.

- n, deneme sayısı 20 oldu.
- p, başarılı olma olasılığı ise 0.6'dır.
- grafik olasılık dağılımını ifade eder. hangi değerlere karşılık hangi olasılık değerlerinin çıkacak olduğunu gösterir.

Dağılım yayıldı ve ortaya gitti. 

### Örnek

bir madeni para 4 kere atılıyor. 2 kere yazı gelmesi olasılığı nedir? p=0.50

- n denemede k başarı elde edilmesiyle ilgileniyor. dolayısıyla bu bir binom dağılımıdır.
- n=4, x=2, p=0.50
- f(2;4,0.50)=(4 2)*0.50^2*(1-0.50)^(4-2)=0.375

# Uygulama

${\displaystyle f(k;n,p)={n \choose k}p^{k}(1-p)^{n-k}}$, k = 0, 1, 2, ...,n


${\displaystyle \operatorname {E} (X)=np}$


${\displaystyle \operatorname {Var} (X)=np(1-p)}$

```python
from scipy.stats import binom
```

bir mecrada reklam veriliyor bu reklam verildiğinde örneğin 100 kişi gördüğünde 1,5 ve 10 tıklama olasılıkları merak ediliyor

```python
p=0.01 # başarılı olma olasılığı
n=100 # deneme sayısı
```

```python
rv=binom(n,p)
```

```python
# pmf: olasılık kütle fonksiyonu
print(rv.pmf(1)) # reklamı 1 kişinin görme olasılığı
print(rv.pmf(5))
print(rv.pmf(10))
```
