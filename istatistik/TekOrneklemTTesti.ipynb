# Tek Örneklem T Testi

Popülasyon ortalaması ile varsayımsal bir değer arasında istatistiksel olarak anlamlı bir farklılık olup olamadığını test etmek için kullanılan parametrik bir testtir.

----

örnek ortalamasına ilişkin test yapmak için kullanılır.

örneklem bölümünde ele aldığımız örneğe geri dönersek 10000 kişilik bir ilçeden 100 kişilik bir örneklem çekmiştik. buna tek örnek denir. bu örnek üzerinden ana kitleye ilişkin bir değerlendirmede bulunmak istiyorsak bu popülasyon ortalamasına karşılık gelir. bu ana kitle ortalamasına ilişkin teorik bir değer ile bir test yapmak istersek bu durumda tek örneklem t testi yapmış oluruz

## Tek Örneklem T Testi: Hipotezler

- H0: μ = 50
- H1: μ ≠ 50

- H0: μ ≤ 50
- H1: μ > 50

- H0: μ ≥ 50
- H1: μ < 50

H0'lar dayanak noktası olarak düşünülmelidir.

## Tek Örneklem T Testi: Test İstatistiği



- Tek örneklem t testini yapmak istediğimizde t veya z dağılımlarını kullanabiliyoruz.

1. Anakütle standart sapması biliniyorsa z istatistiği kullanılır.
2. Anakütle standart sapması bilinmiyorsa ve n > 30 ise z istatistiği kullanılır.
3. Anakütle standart sapması bilinmiyor ve n < 30 ise t istatistiği kullanılır.

- bunlar tek örneklem t testinde, test istatistiğine karar vermek için yapılacak olan seçimdir.

- n(örnek sayısı) büyüdükçe t, z'ye yaklaşır.

- veri bilimi kapsamında genelde elimizdeki n, 30'dan büyük olacağından ele alacağımız örneklerde bu değerler genelde 30'dan büyük olacağından
ve örnek sayısı arttıkça t dağılımı, normal dağılıma yaklaşacağından bu sebeple sürekli t üzerinden düşünülebilir.

- python programlarının içerisinde kullanılan test istatistiği genelde t olmuş olur.

# Tek Örneklem T Testi: Varsayım

Hipotez testlerinde belirli varsayımlarımız olacaktır ve bu varsayımları test etmemiz gerekecektir. Bu noktada, eğer varsayımlarımız sağlanıyorsa ilgili hipotez testini kullanabileceğiz. Ancak, varsayımlar sağlanmıyorsa, uygulama bölümünde bu durumu değerlendirmeli ve parametrik olmayan (non-parametrik) hipotez testlerine geçiş yapmalıyız. Hipotez testlerinin genel adı "parametrik hipotez testleri" olsa da, bazı ön koşullar veya varsayımlar sağlanmadığında bu testleri kullanamayabiliriz. Gerçek dünya verilerinde sıkça karşılaşılan bu durumlar için non-parametrik testleri kullanacağız ve bu konuda daha fazla bilgi edineceğiz.

### Normal Dağılım

----
Hipotez testlerimizde bazı varsayımlarımız olacak ve bizim bu varsayımları test etmemiz gerektiği gerçeği olacak bizler bu varsayımları test ederek eğer varsayım sağlanıyorsa ilgili hipotez testini kullanacağız eğer varsayım sağlanmıyorsa uygulama bölümünde bunu değerlendirmiş olacağız ve NaN parametrik hipotez testlerini ele alacağız. hipotez testi diyerek ele aldığımız bütün hipotez testleri aslında parametrik hipotez testleri fakat varsayılan bazı ön koşullar yani varsayımlar sağlanmadığında bunları kullanamayız. veri bilimi gerçek dünyada NaN parametrik durumlar söz konusu NaN parametrik testleri de yani varsayımlar sağlanmadığında kullanılan testleride göreceğiz.

----

Özetle, t testi, tek bir örneklemin ortalamasıyla ilgili test yapma ihtiyacı doğduğunda kullanılan bir testtir. Örneğin, bir ilçenin yaş ortalamasıyla ilgili bir hipotez testi kurmak istediğimizde bu testi kullanabiliriz. Aynı şekilde, web sitemizde insanların geçirdiği süreyle ilgili bir hipotez testi yapmayı düşünüyorsak, 180 saniye olup olmadığını, 180 saniyeden uzun veya kısa olup olmadığını test etmek isteyebiliriz. Bu tür durumlarda elimizde tek bir örneklem, tek bir grup ve tek bir değişken bulunur. Bu nedenle tek örneklem t testi kullanılır.
