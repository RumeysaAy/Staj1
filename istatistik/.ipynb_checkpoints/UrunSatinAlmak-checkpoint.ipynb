# İş Uygulaması: Ürün Satın Alma Adım Optimizasyonu

- Bir e-ticaret şirketi, internet üzerinden çeşitli ürünler satmaktadır ve bu ürünlerin satış süreci birkaç adımdan oluşmaktadır.
- Bu adımların kullanıcı deneyimi açısından optimize edilmesi ve kullanıcı deneyiminin iyileştirilmesi gibi çabalar şirketin ana hedeflerindendir.
- Kullanıcı deneyimi mümkün olduğunca basit ve hızlı bir şekilde sağlanmaya çalışılırken, ortaya çıkan problemler de çözülmeye çalışılmaktadır.

## Problem:
Sepete ürün ekleme işlemi sonrasında ödeme ekranında 5 adım bulunmaktadır ve bu adımlardan birisi sorgulanmaktadır.

## Detaylar:
- Her adımın hedeflenen süresi 20 saniyedir. 4. adım sorgulanmaktadır.
- Bu durumu test etmek için 100 örnek alınmıştır.
- Örneklem standart sapması 5 saniye, örneklem ortalaması ise 19 saniyedir.

Ürünümüzü kullanan on binlerce müşterimiz bulunmaktadır. Her bir müşteri, satın alma sürecinde belirli bir zaman harcamaktadır. Ancak, bu on binlerce kullanıcının tüm kullanıcı kitlesini tam anlamıyla temsil etmediği bilinmektedir. Yani sistemimize ulaşmayan çok sayıda müşteri bulunmaktadır. Dolayısıyla aslında ana kitleye tam erişim sağlayamıyoruz. Bu nedenle, mevcut müşteriler arasından belirli bir örneklem alınması gerekmektedir. Bu örneklem, 100 veya 1000 gibi değişebilir. Seçilen bu örnekleme dayanarak bir hipotez testi oluşturulabilir ve böylece bu sürenin gerçekten 20 saniye olup olmadığını test etmiş oluruz.

100 örnek çektik ve bu örneklerin ortalaması zaten 19. Bu durumda neyi test ediyoruz derseniz, aslında şunun testini yapıyoruz: Ortaya çıkan bir etkinin sadece şansa dayalı bir sonuç olup olmadığını, istatistiksel olarak güvenilir bir şekilde ispat etmeye çalışıyoruz.

Peki, ortalama zaten 19 ise test etmek niye önemli? İşte burada standart sapmayı da göz önünde bulundurarak, istatistiksel olarak anlamlı bir sonuç olup olmadığını değerlendireceğiz.

**Adım 1**: Hipotezlerin Kurulması ve Yönlerinin Belirlenmesi
- H0: μ = 20 -> Bilinmeyen anakütle ortalaması (μ), yani kişilerin tüm insanların 4. adımda geçirdikleri süre ortalaması 20'dir.
- H1: μ ≠ 20 -> Bilinmeyen anakütle ortalaması (μ), yani kişilerin tüm insanların 4. adımda geçirdikleri süre ortalaması 20 saniyeden farklıdır.

**Adım 2**: Anlamlılık Düzeyinin ve Tablo Değerinin Belirlenmesi
- α = 0.05 -> α/2 = 0.025
- Ztablo için tablo olasılık değeri: 0.5 - 0.025 = 0.475
- Ztablo kritik değer = -1.96 veya +1.96

Anlamlılık düzeyi, yani kabul edilebilir hata miktarı α (alfa) olarak adlandırılır. İki yönlü bir hipotez testi yapacağımız için başlangıçta belirlediğimiz α'yı yani 0.05'i ikiye böleriz. Ancak "büyüktür" ya da "küçüktür" şeklinde bir alternatif hipotezimiz olsaydı, α'yı olduğu gibi kabul edip bu α'ya karşılık gelen tablo değerlerine bakacaktık. Ancak alternatif hipotezimiz iki yönlü olduğundan α'yı ikiye bölüp bunun üzerinden bir Z tablo kritik değeri bulacağız.

Z tablo olasılık değerini bulmak için 0.5'ten 0.025 çıkartarak 0.475 değerine karşılık gelen tablo değerine bakmamız gerekiyor. Buna karşılık -1.96 veya +1.96 olarak gelir.

bu uygulamada ana kitlenin standart sapmasını bilmiyorduk fakat n > 30 olduğundan z dağılımını kullandık. n yeteri kadar büyük olduğundan t dağılımına ilişkin tabloda kullanılabilir. biz burada z'yi ele almış olduk.



- Bu grafiğin tamamı 1'e eşittir.
- Grafiği 0 noktasından ikiye ayırdığımızda sağ tarafı 0.5'e, sol tarafı ise 0.5'e eşittir.
- Çalışmanın başında kabul edilebilir hata miktarı olarak belirlemiş olduğumuz α'yı ikiye böldükten sonra elde ettiğimiz değer, bu tabloda verilen eğrinin altında kalan alanları ifade eder. 
- Bulduğumuz 0.475 değerini tabloda arayarak bu olasılık değerine karşılık gelen z değerlerini bulacağız.
- Tabloda 0.475 olasılık değerini bulduktan sonra buna karşılık gelen kritik değeri yani z karşılığını bulmalıyız. Bu değere karşılık gelen y ve x eksenindeki değerlere baktığımızda 1.96 bulunur.
- Yani, elimizdeki kabul edilebilir hata miktarını kullanarak olasılık değerini hesaplıyoruz. Bu değeri standartlaştırılmış bir tablo üzerinden arıyoruz ve kritik değeri buluyoruz. 1.96 olarak z tablo değerini bulduk
- Güven aralığı bölümünde genellikle %95 güvenirlilikle çalıştığımızda α değerini 1.96 olarak kabul etmemiz gerekiyordu. Çünkü orada da + ve - yönlü bir işlem yapıyor olduğumuzdan dolayı, güven aralığı söz konusu olduğunda 1.96'yı Z tablo değeri olarak kabul etmiştik. İşte o Z tablo değerini bulmanın yolu bu tablodur.

Özetle, 2. adım:
- α=0.05'ti, fakat iki yönlü bir hipotez kurduğumuzdan α'yı ikiye böldük: α/2=0.025
- Bu grafiğin toplam alanı 1'e eşitti. Yarısı da 0.5 idi. Z değerine gitmek için 0.5-0.025=0.475 olasılık değeri bulunur.
- Tabloda, 0.475'e karşılık gelen değer -1.96 ve +1.96 olarak bulunur.

Karşılaştırma yapabilmek adına kritik değerimiz, yani tablo değerimiz 1.96'dır.

**Adım 3**: Test istatistiğinin belirlenmesi ve test istatistiğinin hesaplanması



n=100, standart sapma=5, örnek ortalaması 19 sn

- n>30 olduğundan z'yi kullandık. t'de kullanılabilirdi. çok yakın sonuçlar olacaktır.
- z test istatistiğimizde ilgili değerleri yerine yazdığımızda örneklem ortalaması 19, sınamış olduğumuz hipotezdeki teorik değerimiz 20, çalışmanın standart sapması (ana kitleye ilişkin standart sapma değeri elimizde olmadığından örneklemin standart sapmasını kullandık) 5, gözlem sayısı(örneklemdeki gözlem sayısı) 100'dür. z hesap istatistiği -2 olarak bulundu.
- elimizde bir z tablo değeri var ve bir z hesap değeri var. şimdi bu ikisini karşılaştırarak karar vereceğiz

**Adım 4**: Ztablo ve Zhesap karşılaştırması: Zh > Zt ya da -Zh < -Zt ise H0 Red

Zhesap=-2.00 < Ztablo=-1.96 olduğu için H0 reddedilir.



- Zhesap - çıktığından, Ztabloyu da - olarak değerlendireceğiz.
- yeşil alanlar karar noktalarını ifade ediyor. tablo üzerinden bulmuş olduğumuz değer - 1.96 hesaplamış olduğumuz değer ise -2 değeridir. -2 değeri H0'ın red bölgesi olan yeşil bölgeye düşmektedir. bu yüzden H0 hipotezi reddedilir.

eğer elimizde Zhesap + olsaydı. bu durumda sağ taraftaki yeşil alanla karşılaştırıp + 1.96 ile karşılaştırırdık. buna göre yorum yapacaktık.

Zhesap değeri buradaki red bölgelerine, yeşil alanlara düşmeseydi bu durumda H0 hipotezini reddedmek için yeterli kanıt bulunamamıştır. istatistiki olarak anlamlı bir sonuç elde edilememiştir.

**Adım 5**: Yorum

- H0: μ = 20 -> Bilinmeyen anakütle ortalaması (μ), yani kişilerin tüm insanların 4. adımda geçirdikleri süre ortalaması 20'dir.
- H1: μ ≠ 20 -> Bilinmeyen anakütle ortalaması (μ), yani kişilerin tüm insanların 4. adımda geçirdikleri süre ortalaması 20 saniyeden farklıdır.

4. adımda geçirilen sürenin 20 saniye olduğunu iddia eden H0 hipotezi reddedilmiştir. Buna göre kullanıcılar istatistiksel olarak %95 güvenilirlik ile 4. adımda 20 saniyeden farklı zaman geçirmektedir.

ana kitle dediğimiz bütün insanlar web sitemizin 4. adımında 20 saniye zaman geçirmiyor. bunun üzerine gerekli kararları artık alabiliriz.

eğer H0'ı reddedemeseydik bu durumda kullanıcılarımızı web sitemizde 4. adımda yaklaşık 20 saniye zaman geçirdiğini kabul ederdik ve bu noktada karar alabilirdik.

bu şekilde bütün kullanıcı deneyimini hipotez testleri üzerine oturtturursak elde edilebilecek kararlar tamamen bilimsel olur.

burada kolaylık olması açısından normallik varsayımı incelenmemiştir. uygulamada elimizdeki tek örneklemin normallik varsayımını sağlayıp sağlamadığı kontrol edilecektir. eğer sağlanmıyorsa NaN parametrik t testi yapılmalıdır.
