# ki rneklem Oran Testi: Uygulama

```python
import numpy as np
import pandas as pd
# test
from statsmodels.stats.proportion import proportions_ztest
```

```python
# tklama
basari_sayisi = np.array([300, 250])
```

```python
# gzlem birimi says
gozlem_sayilari= np.array([1000, 1100])
```

```python
proportions_ztest(count=basari_sayisi, nobs=gozlem_sayilari)
```

- p-value < 0.05 salandndan H0 hipotezi reddedilir.
- "eittir" ve "eit deildir" zerinden yorum yaparsak
    - H0: P1 = P2
- H0: web sitemizde tasarlam olduumuz 1. buton renginin dnm oranyla 2. buton renginin dnm oran birbirlerine eittir. ikisi arasnda anlaml bir farkllk yoktur.
- H0 hipotezi, p-value deerimize gre reddedildi. yani hazrlanan buton renkleri incelendiinde bu butonlarn kullanc davranlarna etkisi farkldr ve bu farkllk yeil butonun lehine olacaktr.
- yani iki rnek orannn arasnda istatistiksel olarak anlaml farkllk vardr.
