# İş Uygulaması: Kullanıcı Arayüzü Deneyi (AB Testi)

- Kullanıcı arayüzü deneyim/tasarımı optimizasyonu
- genellikle bir web sitesinin tasarımında ortaya koyulabilecek olası tasarımlardan hangisinin tercih edilmesi gerektiğine karar vermek için kullanılmaktadır.
- iki tane aynı tasarımlı sitemiz var. bu sitelerden birisinin butonu kırmızı, diğerinin butonu yeşil 
- işte bu butonların hangisinin tasarımda kullanılması gerektiğine karar vermek için AB Testi dediğimiz iki örneklem oran testi kullanılmaktadır. 
- iş uygulaması kapsamında iki orana ilişkin gerçek hayat dönüşüm oranı testi olacak.

### Kırmızı Buton mu Yeşil Buton mu?

- Elimizde iki tane buton var. bu butonlardan hangisinin kullanılması gerektiğine karar vereceğiz.



- pythondaki testlerimiz genelde python içerisindeki fonksiyon yapısından dolayı hipotez "eşit değildir" şeklinde olacak dolayısıyla "eşit değildir" şeklinde ilgili hipotezi kurup sonuçları aldıktan sonra zaten yapısını bilmiş olduğumuz oranlar ya da ortalamalar üzerinden hangisinin lehine yorum yapmamız gerektiğini değerlendirmiş olacağız.

- yeşil buton
    - 1000 görüntülenme
    - 300 tıklama
    
    
- kırmızı buton
    - 1100 görüntülenme
    - 250 tıklama

# İki Örneklem Oran Testi: Uygulama

```python
import numpy as np
import pandas as pd
# test
from statsmodels.stats.proportion import proportions_ztest
```

```python
# tıklama
basari_sayisi = np.array([300, 250])
```

```python
# gözlem birimi sayısı
gozlem_sayilari= np.array([1000, 1100])
```

```python
proportions_ztest(count=basari_sayisi, nobs=gozlem_sayilari)
```

- p-value < 0.05 sağlandığından H0 hipotezi reddedilir.
- "eşittir" ve "eşit değildir" üzerinden yorum yaparsak
    - H0: P1 = P2
- H0: web sitemizde tasarlamış olduğumuz 1. buton renginin dönüşüm oranıyla 2. buton renginin dönüşüm oranı birbirlerine eşittir. ikisi arasında anlamlı bir farklılık yoktur.
- H0 hipotezi, p-value değerimize göre reddedildi. yani hazırlanan buton renkleri incelendiğinde bu butonların kullanıcı davranışlarına etkisi farklıdır ve bu farklılık yeşil butonun lehine olacaktır.
- yani iki örnek oranının arasında istatistiksel olarak anlamlı farklılık vardır.
