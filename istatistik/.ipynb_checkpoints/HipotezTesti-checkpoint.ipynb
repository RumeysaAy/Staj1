# Hipotez Testi

Bir inanışı(bir savı, bir tahmini vs) test etmek için kullanılan istatistiksel bir tekniktir.

Bir doktor, bir hastalık için ilaç bulsun ve bu ilacın etkili olduğunu düşünüyor. Bu ilacın etkili olduğunu düşünmek konusu subjektif bir konudur. Bunun bilimsel bir test ile ortaya konması gerekir. Yani bu doktor, 'Ben bir ilaç buldum ve bu ilaç çok etkili, alın bunu kullanın' diyemez. Bu ilacın etkisinin bilimsel olarak kanıtlanması lazım, ispat edilmelidir.

İspat, ortaya çıkan etkinin şans eseri oluşma ihtimali göz önünde bulundurulduğu halde etkinin ispat edilmesi gerektiği anlamına gelir.

Doktor ilaç buldu, bir grup hastayı alırız. Bu gruba hiçbir ilaç verilmez. Bir grup hasta daha alırız, bu gruba doktorun ilacını veririz. Bu iki grubun kan değerlerine bakılır. İlaç etkiliyse bunun kan ölçümlerine yansıması gerekir. Şans eseri bir etki de ortaya çıkmış olabilir. Hipotez testleri, bu şans eseri ortaya çıkma durumunu da göz önünde bulundurarak şansa yer vermeyecek şekilde bize ilgilendiğimiz konuda bir ispat etme imkanı sağlar.

Hipotez testleri, şansa yer vermeyecek şekilde bize ilgilendiğimiz konuda bir ispat etme imkanı sağlar.

Hipotez testi, bize bilimsel bir dayanak sağlar.

Hipotez testleri, bizim için karar verme noktasında bilimsel bir teknik olarak kullanacağımız bir tekniktir.

# Hipotezler ve Türleri

Hipotez testi yapabilmek için teorik olarak bir gösterim sunmamız gerekmektedir. Bu gösterim üç farklı çeşitte gerçekleşebilir.

Hipotezler H0 ve H1 olarak ikiye ayrılır:

- H0: μ = 50
- H1: μ ≠ 50

0 (sıfır) hipotezi ve alternatif hipotez olarak adlandırılır.

Belirli değerler verilerek parametrelere dayalı olarak oluşturulan hipotezlere H0, null veya 0 (sıfır) hipotezi denir. Alternatif hipotez ise H1 olarak ifade edilir.

## Yönlü Hipotezler

- H0: μ ≤ 50
- H1: μ > 50

- H0: μ ≥ 50
- H1: μ < 50

Teorik olarak sınayacağımız ve dayanacağımız hipotez aslında sabit gibi kabul ettiğimiz H0 hipotezidir. Buna dayalı olarak iddiamızı test etmeye çalışacağız.

Örneğin, örneklem bölümünde 10000 gözlemli bir ilçeden 100 tane gözlem birimi çektik ve bunun üzerinden bir ortalama hesapladık. Bu durumu hipotez testlerine uyarladığımızda hipotezimiz şu şekilde olabilir: Ana kitle ortalaması ($μ$) burada yaşları ifade ediyorsa,

Ana kitle ortalaması (yaşlar):
- H0: Bu ilçenin yaşları 50'dir. (H0: μ = 50)
- H1: 50'den farklıdır. (H1: μ ≠ 50)

Bunu hipoteze oturtturduk ve 100 kişilik bir örneklem çekip ortalamasını hesaplayarak gerçekte 50 olup olmadığını sınayabilirdik.

İki yönlü olarak "H1: μ ≠ 50" veya "H1: μ > 50" ve "H1: μ < 50" şeklinde 3 farklı şekilde kurabilirdik.

Tek bir örnekleme için hipotez testi yaparken örneklem sonuçlarını ilerleyen bölümlerde öğreneceğimiz yöntemlerle işlemlere tabii tutarak bilimsel olarak sınayabiliriz.

# Hata Tipleri

Önceki bölümde kurmuş olduğumuz hipotezleri test ettiğimizde, yani gerçekleştirdiğimizde, hatalar ortaya çıkan sonuçlarla ilgili değerlendirmelerdir.



Gerçek durumu, yani sol eksen H0'in doğru ya da yanlış olmasını ifade ediyor. H0: μ = 50
H0 hipotezine göre H1 hipotezini de kurduk. "H1: μ ≠ 50", "H1: μ > 50" ya da "H1: μ < 50" bunlardan birisini kurduk ve testimizi gerçekleştirdik.

Hipotez Testi Sonucu Verilen Karar bölümünde, bu yapmış olduğumuz testin sonucunda verdiğimiz karar olacak. Karar vermek, kurmuş olduğumuz H0 hipotezini reddedeceğiz ya da reddedemeyeceğiz demektir.

H0'ın gerçek durumu var. Bu gerçek duruma karşılık, bunu reddedip reddedememe durumumuz var.

H0 gerçekte doğruyken eğer H0'ı reddetmezsek, yani H0'ı kabul edersek, bu durumda doğru bir karar vermiş oluruz. (Doğru Karar) Buna güven düzeyi denir. (1-alfa)

H0 gerçekte doğruyken eğer H0'ı reddedersek, hatalı bir iş yapmış oluruz. Buna 1. tip hata denir. Alfa hatası denir.

H0 gerçekte yanlış iken eğer H0'ı reddetmezsek, yani yanlış olanı doğru kabul edersek, bu durumda yine hata yapmış oluruz ve buna 2. tip hata denir. Beta hatasıdır.

H0 gerçekte yanlış iken eğer H0'ı reddedersek, yani yanlış olanı reddedersek, bu durumda yine doğru bir iş yapmış oluruz. (Doğru Karar) Bu da "Testin Gücü"nü gösterir. (1-beta)

Daima bu hatalardan birini yapma ihtimali vardır. Bu ihtimalleri risk derecesini belirleyerek azaltmak isteriz. Yine bir belirsizlik var ve biz belirsizlikle ilgili yorumlar yapmaya çalışıyoruz. Burada ilgilenecek olduğumuz 1. tip hata dediğimiz alfa hatası...

H0 = 50 mi, değil mi, bilmiyoruz. Bir test yapacağız ve belirsizliği azaltmaya çalışıyoruz.
