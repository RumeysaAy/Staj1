- Veri bilimciler olarak amacımız belirsizlik altında karar vermektir. Buna yönelik olarak bazı tekniklerle belirsizliği azaltmaya çalışıyoruz. Bu belirsizliği azaltmaya çalıştığımız yaklaşımlardan biriside olasılık dağılımlarıdır. 
- olasılık dağılımları bize bazı değerlerin ne şekilde hangi olasılıklarla gerçekleşebileceğine yönelik fikir verir. 
- örneğin satışlarımızın şu değerden fazla olma olasılığı, 5 günde 3 defa şöyle bir olayın gerçekleşmesi olasılığı normalde bunlarla ilgili bilgi yoktu ama olasılık teorisi aracılığıyla bu konulardaki belirsizlik azaltılmaya çalışılır. bilgi sunulur. 
- amacımız olasılık değerleri hesaplamak. bunları da bazı kesikli ve sürekli değişkenler çerçevesinde gerçekleştirebiliyor olacağız.

# Kesikli Olasılık Dağılımları

## Bernoulli Dağılımı

Başarılı-başarısız, olumlu-olumsuz, 1-0 şeklindeki iki sonuçlu olaylarla ilgilenildiğinde kullanılan kesikli olasılık dağılımıdır.

Bir kategorik değişken olarak düşünürsek, nominaldir.

Burada hepsini sayılar cinsinden ifade ettiğimizden dolayı, kesikli ya da sürekli şeklinde ifade edeceğiz.

kesikli rassal değişkenlerden birisi bernoulli 1-0, başarılı-başarısız...

1-0, başarılı-başarısız vb. şeklinde, sanki bir kategorik değişken gibi sadece belirli yapıları işaret eden değişkenler olduğunda, bunlara kesikli deriz.

Sürekli ise 1 ile 2 arasında birçok değer alınabilir; 1.2, 1.3, 1.4 vb. Burada sürekli rassal değişkenler kullanılır.

## Bernoulli Dağılımı

p, ilgilendiğimiz bu olaydaki bir sınıfın gerçekleşme olasılığını ifade eder. İki olayla ilgilendiğimizi ifade etmiştik. Örneğin, yazı-tura deneyinde yazı gelme olasılığı %50 ise, p 0.5'tir; yani %50'yi ifade eder. x kesikli rassal değişkenin alacağı değerdir. Bu Bernoulli dağılımı kapsamında 0 ya da 1 olabilir. Yazı için 1, tura için 0 şeklinde. Bu fonksiyon bize ilgili değerleri yerine yazdığımızda olasılık hesabı yapacak ve yine bütün dağılımlarla beraber beklenen değer (E(X)) ifadesi ile varyans (Var(X)) ifadelerini vermiş olacağız.

elimizde eğer bir dağılım varsa bir değişken gibi bunu düşüneceğiz ve bunlarında bir beklenen değeri yani ortalaması beklenen değer ortalamaya karşılık geliyor yani merkezi eğilimi ifade eder varyans ise dağılımını yayılımını ifade eder. elimizdeki dağılımlarla ilgili beklenen değer(E(X)) ve varyans (Var(X)) varsa bunlar dağılımın merkezini ve o merkez etrafındaki dağılımını ifade ediyor olacak.

- bernolli dağılımının beklenen değeri p, varyansı ise p*(1-p) yani ilgili olayın gerçekleşme olasılığı çarpı gerçekleşmemesi olasılığı varyansıdır.
- ilgili olayın gerçekleşmesi olasılığı, beklenen değeri ifade eder.

Bernoulli kesikli rassal değişkeninin alabileceği değerler 1 ve 0'dır.
Bu değişken 0 değerini aldığında olasılık değeri 1-p,
1 değerini aldığında ise olasılık değeri p olacaktır.
İlgilendiğimiz ana sınıfı 1 olarak kabul edip bu rassal değişkenin x değeri 1 olduğunda olasılık p,
0 olduğunda olasılık ise 1-p değerine karşılık gelir.

# Uygulama

${\displaystyle f(k;p)=p^{k}(1-p)^{1-k}\quad {\text{for }}k\in \{0,1\}}$


${\displaystyle \operatorname {E} \left(X\right)=p}$

$\operatorname {Var} [X]=pq=p(1-p)$

```python
# olasılık dağılımları ile ilgili işlemler için scipy kütüphanesi kullanırız
# scipy kütüphanesinin içerisinden stats'ın bernoulli fonksiyonunu import ettik
from scipy.stats import bernoulli
```

```python
# p, iki sonuçlu bir olayı ifade eder. yazı-tura deneyinde tura gelme olasılığı 0.6 olsun
p= 0.6
```

```python
# bernoulli dağılımı ile olasılık hesaplayalım
rv= bernoulli(p)
```

```python
# k=1 ilgilendiğimiz olay True olma olasılığıdır. bunu 0.6(%60) olarak vermiş olduk p= 0.6
rv.pmf(k=1) # pmf: olasılık kütle fonksiyonu
```

```python
# k=0 ilgilendiğimiz olay False olma olasılığıdır. 0.4(%40) yazı
rv.pmf(k=0)
```
