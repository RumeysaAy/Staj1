# Korelasyon Analizi

- Korelasyon, ilişki analizi demektir.
- Değişkenler arasındaki ilişki, bu ilişkinin yönü ve şiddeti ile ilgili bilgiler sağlayan istatistiksel bir yöntemdir.

## Korelasyon Türleri ve Yönleri



iki değişken arasında
- birinci grafikte görüldüğü bir şekilde bir yapı sergileniyorsa bu durumda değişkenler arasında "Mükemmel Pozitif Korelasyon" vardır. Bu "1" ile gösterilir.
- Mükemmel Negatif Korelasyon" ise "-1" ile gösterilir.
- Değişkenler arasındaki ilişki bu skalada değişmiş olur.
- 0.9 ile 0.5 arasında bir ilişki gözlemlendiğinde bu ilişkilerin anlamlı olduğu, dikkate değer olduğunu düşünürüz. bunlardan düşük olan ilişkileri ise çok anlamlı değil şeklinde değerlendiririz.

- Ürün satışı ile ürün reklam harcaması değişkenlerini düşünelim. Ürün reklam harcamaları arttıkça ürün satışlarının arttığı gözlemlenmektedir. bu iki değişken arasında pozitif yönlü bir korelasyon, ilişki olacaktır.

- Araç kilometresiyle fiyatı düşündüğümüzde aracın kilometresi arttığında fiyatların düşünmesini bekleriz. bu negatif yönlü bir korelasyona örnektir.

Pearson korelasyon katsayısı 0'dan büyükse pozitif yönlü ilişki, 0'dan küçükse negatif yönlü ilişki var. 1'e veya -1'e yaklaştıkça ilişkinin şiddeti artmaktadır. 0'a geldikçe ilişkinin şiddeti düşmektedir. 

## Korelasyon Analizi: Hipotezler

### Korelasyonun Anlamlılığının Testi
Korelasyon katsayısını bulduk ama bu anlamlı mıdır?

- H0: Ro(ρ) = 0
- H1: Ro(ρ) ≠ 0

- H0: değişkenler arasında korelasyon yoktur, anlamına gelir.
- H1: değişkenler arasında anlamlı bir ilişki vardır, anlamına gelir.

yapılacak olan hipotez testinden elde edilen p-value değerinin 0.05'ten küçük olup H0'ı reddedmemize sebep olmasını bekleriz.

## İş Uygulaması: Bahşiş ile Ödenen Hesap Arasındaki İlişkinin İncelenmesi

Bahşiş ile ödenen hesap arasında korelasyon var mı?

- H0: Ro(ρ) = 0 ödenen hesap ile bahşiş arasında korelasyon yoktur.
- H1: Ro(ρ) ≠ 0 vardır.

- Bahşiş veri seti:
    - total_bill: yemeğin toplam fiyatı (bahşiş ve vergi dahil)
    - tip: bahşiş
    - sex: ücreti ödeyen kişinin cinsiyeti (0=male, 1=female)
    - smoker: grupta sigara içen var mı? (0=No, 1=Yes)
    - day: gün (3=Thur, 4=Fri, 5=Sat, 6=Sun)
    - time: ne zaman? (0=Day, 1=Night)
    - size: grupta kaç kişi var?

```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
df=tips.copy()
df.head()
```

Bahşiş ile yemeğin toplam fiyatı arasındaki ilişkiyi merak ediyoruz. Fakat burada bir problem var çünkü veri setinin hikayesi incelendiğinde ödenen toplam fiyata odaklanırsak bahşiş ve verginin dahil olduğu görülür. bu korelasyon incelemelerinde özellikle birbirleriyle doğrudan ilişkisi olduğu bilinen senaryolarda değişkenlerin birbirlerine olan etkileri incelenmeden önce eğer varsa değişkenlerin birbirlerine olan etkilerinin arındırılması gerekmektedir.

toplam fiyat ile bahşiş arasındaki korelasyona bakacağız fakat ama toplam fiyatın içerisinde bahşişde vardır. bunu düzeltelim:

```python
# toplam hesaptan bahşişi çıkaralım:
df["total_bill"] = df["total_bill"] - df["tip"]
df.head()
```

```python
# korelasyon grafiği oluşturalım:
df.plot.scatter("total_bill","tip");
```

Toplam hesap arttıkça bahşiş arttığı görülür.

## Korelasyon Varsayım Kontrolü

### Normallik Varsayımı

Elimizdeki iki değişken içinde normallik varsayımının sağlanması gerekiyordu.

```python
from scipy.stats import shapiro
```

```python
test_istatistigi, pvalue = shapiro(df["tip"])
print("Test İstatistiği = %.4f, p-değeri = %.4f" % (test_istatistigi, pvalue))
```

```python
test_istatistigi, pvalue = shapiro(df["total_bill"])
print("Test İstatistiği = %.4f, p-değeri = %.4f" % (test_istatistigi, pvalue))
```

- H0: örnek dağılımıyla teorik normal dağılım arasında istatistiksel bir anlamlı fark yoktur.
- pvalue < 0.05 sağlandığından H0 reddedilmiştir. normallik varsayımı sağlanmamaktadır.
- örnek dağılımıyla teorik normal dağılım arasında istatistiksel bir anlamlı fark vardır. fark varsa örnek dağılım ile teorik normal dağılım birbirine benzemiyor demektir.
- çok önemli olan normallik varsayımı sağlanmamaktadır.

- Normallik varsayımının sağlandığını varsayıp, parametrik testimizi yapalım.
- sağlamadığı için nonparametrik testimizi yapalım.

# Korelasyon Katsayısı Hipotez Testi

## Korelasyon Katsayısı

```python
## Parametrik Yöntemle: Pearson
# Pearson korelasyon katsayısı
df["tip"].corr(df["total_bill"])
```

- bu corr fonksiyonu ön tanımlı olarak pearson katsayısını bize verir.
- pearson korelasyon katsayısı ancak değişkenler için normallik varsayımı sağlandığında hesaplanabilir.
- fakat biz gözlemlediğimizde bu sağlanmıyordu.
- bu sebeple corr içerisine bir argüman aracılığıyla spearman korelasyon katsayısını girip ilgili korelasyonu alabiliriz.

```python
## Nonparametrik Yöntemle: Spearman
# Spearman korelasyon katsayısı
df["tip"].corr(df["total_bill"], method="spearman")
```

- Değişkenlerin arasında pozitif yönlü bir ilişki olduğu görülür.
- İlişkinin yönü orta şiddetli 0.50 civarinda
- Bu ilişki anlamlı mı?

## Korelasyonunu Anlamlılığının Testi

### pearson korelasyon katsayısı için

```python
from scipy.stats.stats import pearsonr
```

```python
test_istatistigi, pvalue = pearsonr(df["tip"], df["total_bill"])
print("Korelasyon katsayısı = %.4f, p-değeri = %.4f" % (test_istatistigi, pvalue))
```

- H0: değişkenler arasında anlamlı bir ilişki yoktur
- pvalue < 0.05 sağlandı. H0, reddedilir.
- değişkenler arasında anlamlı bir ilişki, korelasyon vardır.

## Nonparametrik Korelasyon Testi, Hipotez Testi

Pearson korelasyon katsayısı için korelasyonun anlamlılığının testini gerçeleştirmiştik fakat varsayımları incelediğimizde varsayımların sağlanmadığını gözlemledik. **varsayımlar _sağlanmadığında_** **spearman** ya da **kendal** ismi verilen 2 tane **nonparametrik** yöntemimiz var.

### Spearman korelasyon katsayısı için

```python
from scipy.stats import stats
```

```python
stats.spearmanr(df["tip"], df["total_bill"])
```

- H0: değişkenler arasında ilişki yoktur.
- pvalue < 0.05 olduğundan H0 reddedilir.
- değişkenler arasında ilişki vardır. bu orta şiddetlidir ve bu ilişki anlamlıdır.

```python
test_istatistigi, pvalue = stats.spearmanr(df["tip"], df["total_bill"])
print("Korelasyon katsayısı = %.4f, p-değeri = %.4f" % (test_istatistigi, pvalue))
```

- pvalue değeri, spearman'a ilişkin anlamlılığın test edildiği p-value değeridir.

iki değişken arasında ilişki olup olmadığı, ilişkinin anlamlı olup olmadığını, bu ilişkinin yönünü ve şiddetini merak ediyorduk buna yönelik olarak 2 değişkenin arasındaki ilişki korelasyon testleriyle test etmeye verdik. fakat öncesinde varsayımları inceledik.
- varsayım sağlanıyorsa pearson,
- varsayım sağlanmıyorsa spearman korelasyon katsayısını kullanacağız.



buna yönelik olarak ilgili korelasyon katsayısını elde ettiğimizde bu korelasyon katsayısına ilişkin hipotez testleri gerçekleştirdiğimizde görünen o ki 
- H0: değişkenleri arasında anlamlı bir korelasyon yoktur.
H0 hipotezi, p-value değeri alfa(0.05)'dan küçük olduğundan reddedilir. yani değişkenlerin arasında istatistiksel olarak anlamlı bir ilişki vardır. bu ilişki pozitif yönlüdür, 0.5 civarinda ve orta şiddete sahiptir.


- çalışmanın başında kabul ettiğimiz, kabul edilebilir hata miktarı olan alfa, 0.05'dir.

### Kendall Tau korelasyon katsayısı için

```python
test_istatistigi, pvalue = stats.kendalltau(df["tip"], df["total_bill"])
print("Korelasyon katsayısı = %.4f, p-değeri = %.4f" % (test_istatistigi, pvalue))
```

kendalltau ile spearmanr kıyasladığımızda en güvenilir spearmanr'dır.
