# Normal Dağılım

Normal dağıldığı bilinen sürekli rassal değişkenler için olasılık hesaplaması için kullanılır.

- elimizde bir sürekli değişken var. bu sürekli değişkenin normal dağılıma sahip olduğu bilgisini edindik. ve bir olasılık değeri hesaplamak istiyoruz.

- f(x | mü, sigma^2) buradaki
    - mü: ortalamayı ifade eder. beklenen değerdir.
    - x: rassal değişkenin değerini ifade eder.
    - sigma^2: varyansı ifade eder.


- normal dağılım fonksiyonunu sürekli değişken kavramı adı altında yani aralık anlamında bazı olasılık değerleri hesaplamak için kullandığımızda bunu bir integral işlemiyle gerçekleştiririz.

- önceki dağılımlarda dağılımları kullanırken olasılık hesabı yapmak istediğimizde kesiklilik durumu olduğundan dolayı değerler ve bu değerlere kerşılık gelen olasılıklar yer almıştı. dağılımlarda burada ise elimizdeki değişken sürekli olduğundan bir hesaplama, olasılık işlemi yapmak için bir integral ile alan hesabı yapmamız gerekli.

# İş Uygulaması: Ürün Satış Olasılıklarının Hesaplanması

## Problem:
Bir yatırım/toplantı öncesinde gelecek ay ile ilgili satışların belirli değerlerde gerçekleşmesi olasılıkları belirlenmek isteniyor.


## Detaylar:
    - Dağılımın normal olduğu biliniyor
    - Aylık ortalama satış sayısı 80k, standart sapması 5k
    
    
## Soru:
90k'dan fazla satış yapma olasılığı nedir?

Ortalama 80k olarak verilmiş. Standart sapma ise 5k olarak verilmiş. Standart sapma, ortalama etrafındaki yayılımı/dağılımı gösterir.

Standart sapmayı ortalamanın üzerine eklersek 85k olur. Çıkarırsak 75k olur. Zaten ortalamanın etrafında 85k ve 75k gibi değerlerin yer almasını bekleriz.

85k ve 75k aralığının içerisinde veya bu aralığa yakın olan değerlerin bu olasılığının yüksek çıkmasını bekleriz. 90k, 85k'dan fazla olduğundan, bunun olasılığı düşecektir.

Normal dağılıma göre çözersek, ilgili değerler fonksiyonda yerine yazıldığında:
- P(X > 90) = 0.0228
- P(X > 70) = 0.9772
- P(X < 73) = 0.0808
- P(85<X<90) = 0.1359



Burada integral işlemi yapılır. Bu integral işlemlerinde üst sınır ve alt sınırlar hesaplanır; buna göre bu olasılık değerlerine ulaşılır.

-----

Eğer standart sapma, ortalama değerden küçükse, bu genellikle verilerin ortalama etrafında daha sıkı bir şekilde gruplandığını ve daha az yayıldığını gösterir. Yani verilerin genel olarak daha homojen veya dengeli olduğunu ifade edebilirsiniz.

standart sapma ne kadar küçükse, veri noktalarının ortalama değere daha yakın olduğu ve bu nedenle genel dağılımın daha sıkı olduğu söylenebilir. Bu durum, veri kümesinin istatistiksel olarak daha istikrarlı veya tahmin edilebilir olduğunu gösterebilir.

**neden 85k ve 75k aralığının içerisinde veya bu aralığa yakın olan değerlerin bu olasılığının yüksek çıkmasını bekleriz ?**

Bu beklenti, istatistiksel dağılımların genel özelliklerinden kaynaklanır. Normal dağılım (veya Gauss dağılımı) gibi birçok istatistiksel dağılım, merkez etrafında simetrik bir şekilde yayılır. Bu dağılımlarda, merkezi değer (ortalama) etrafında daha fazla veri bulunurken, dağılımın uç değerlere doğru seyri giderek azalır. Bu tür dağılımlarda standart sapma, dağılımın genişliği ve yayılım derecesini ifade eder.

Bu nedenle, ortalama etrafındaki belirli bir aralığa (örneğin, ortalama ± 1 standart sapma) denk gelen değerlerin, dağılımın genişliği nedeniyle daha yüksek bir olasılığa sahip olması beklenir. Çünkü bu aralık, dağılımın merkezi ve daha yoğun olan bölgelerini içerir. Daha uzak değerlere gittikçe, dağılımın yoğunluğu azalır ve bu nedenle bu değerlerin gerçekleşme olasılığı daha düşer.

Sonuç olarak, istatistiksel dağılımların simetrisi ve normal dağılımlara yakın davranışı nedeniyle, ortalama etrafındaki belirli aralığa yakın değerlerin daha yüksek bir olasılığa sahip olması beklenir. Bu prensip, istatistiksel tahminler ve analizlerde sıkça kullanılan bir yaklaşımdır.

# Normal Dağılım

${\displaystyle f(x\mid \mu ,\sigma ^{2})={\frac {1}{\sqrt {2\pi \sigma ^{2}}}}e^{-{\frac {(x-\mu )^{2}}{2\sigma ^{2}}}}}$

* $\mu$  ortalama ya da dağılımın beklenen değeri

* ${\displaystyle \sigma }$  standart sapma

* ${\displaystyle \sigma ^{2}}$ varyans

```python
from scipy.stats import norm
```

```python
# 90'dan fazla olması

1-norm.cdf(90, 80, 5) # cdf: kümülatif yoğunluk fonksiyonu 
# cdf, kesiklilerden farklı olarak bir alan taramış olacağız.
# bu kümülatif bir yoğunluk olarak değerlendiriliyor.
```

- 1-norm.cdf(90, 80, 5) # cdf: kümülatif yoğunluk fonksiyonu 

1 den çıkarma işlemini gerçekleştirmemizin sebebi normal eğri yapısının altında kalan alan 1'e eşit bize ise belirli bir integral kısmı lazımdı. bunun için cdf fonksiyonunu kullanarak 90dan yukarısının olasılığını hesaplayabilmek için 1den çıkarma işlemini gerçekleştirdik cdf(hesaplamak istediğimiz değer, ortalama, standart sapma)

Bir olasılık dağılımı altındaki belirli bir aralığın olasılığını hesaplamak için kümülatif yoğunluk fonksiyonu (CDF) kullanılır. Ancak, belirttiğiniz örnek kodun dilini anlamaya çalışıyorum ve kullanmış olduğunuz dil büyük olasılıkla bir programlama dilidir. Özellikle `1-norm.cdf(90, 80, 5)` ifadesi, Python dilinde scipy kütüphanesinin norm fonksiyonu tarafından sağlanan bir CDF hesaplama örneğini temsil ediyor gibi görünüyor.

Bu ifade, bir normal dağılımın (belli bir ortalama ve standart sapmaya sahip) altında kalan alanın belirli bir noktadan (90) yukarısının olasılığını hesaplamaktadır. Normal dağılımda CDF, belirli bir değerin altında kalan alanı temsil eder. Ancak, belirli bir aralığın olasılığını hesaplamak istediğinizde, bu aralığın dışında kalan kısmın olasılığını bulmak için 1'den CDF'yi çıkarırsınız. Bu şekilde, istenen aralığın olasılığını elde edebilirsiniz.

Yani, `1 - norm.cdf(90, 80, 5)` ifadesi, 90 değerinin üzerinde kalan alanın (olasılığın) hesaplanmasını sağlar. `norm.cdf(90, 80, 5)` ifadesi normal dağılımın CDF fonksiyonu ile 90 değerinin altında kalan alanın olasılığını hesaplar, ardından 1'den bu olasılığı çıkararak 90 değerinin üzerinde kalan alanın olasılığını bulmuş olursunuz.

```python
# 70'den fazla olması
1-norm.cdf(70, 80, 5)
# ortalama 80 olduğundan 70'den fazla olma olasılığının yüksek olması beklenir.
```

```python
# 73'den az olması
norm.cdf(73, 80, 5)
```

```python
# 85 ile 90 arasında olması
norm.cdf(90, 80, 5)-norm.cdf(85, 80, 5)
```
