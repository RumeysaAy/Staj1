# İş Uygulaması: Dönüşüm Oranı Testi

Dönüşüm oranı, pazarlama dünyasında ve e-ticaret alanında oldukça önemli bir kavramdır. Örneğin, bir web sitesinde bir ürünü sergilediğimizde, bu ürünü 100 kişi görüyorsa ve 1 kişi satın alıyorsa, bu durumda ürünün dönüşüm oranı 1/100'dür, yani 0.01'dir. Bu bir reklam da olabilirdi.

dönüşüm oranı = (ilgilenilen olay) / (o olayı gerçekleştirmek üzere o duruma mağruz kalan kişi sayısı)

**Problem**:
Bir yazılım ile bir mecrada reklam verilmiş ve bu reklama ilişkin yazılım tarafından 0.125 dönüşüm oranı elde edildiği ifade edilmiş. Fakat bu durum kontrol edilmek isteniyor. Çünkü bu yüksek bir oran ve gelirler incelendiğinde örtüşmüyor.

**Detaylar**:
- 500 kişi dış mecrada reklamlara tıklamış, 40 tanesi sitemize gelip alışveriş yapmış
- Örnek üzerinden elde edilen dönüşüm oranı: 40/500=0.08

0.125 oranını test edeceğiz.

H0: P = 0.125
H1: P ≠ 0.125
