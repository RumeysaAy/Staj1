# İş Uygulaması: ML Modelinin Başarı Testi

Bir makine öğrenmesi modeli geliştirilmiş ve bu model canlı sisteme entegre edilmiş. makine öğrenmesi modelinin başarısı AB testi ile test edilmek isteniyor. A, eski sistemi ifade eder. B, makine öğrenmesi modeliyle önerilerde bulunan yeni sistemi ifade ediyor.

- **Problem**:
Bir ML projesine yatırım yapılmış. Ürettiği tahminler neticesinde oluşan gelir ile eski sistemin ürettiği gelirler karşılaştırılıp anlamlı farklılık olup olmadığı test edilmek isteniyor.

- **Detaylar**:
    - Model geliştirilmiş ve web sitesine entegre edilmiş.
    - Site kullanıcıları belirli bir kurala göre ikiye bölünmüş olsun.
    - bu kullanıcıların bir grubuna eski sisteme göre ürün gösterilirken, B grubuna yani diğer gruba da yeni sisteme göre ürünler gösterilir.
    - A grubu eski B grubu yeni sistem.
    - Gelir anlamında anlamlı bir iş yapılıp yapılmadığı test edilmek isteniyor.
    
- **Hipotezlerimiz**:
**ML modeli anlamlı farklılık oluşturabildi mi?**

- H0: μ1 = μ2
- H1: μ1 ≠ μ2

H0: eski sistemin gösterilerine göre insanlar ürün aldığında belirli bir süre zaarfında ortaya çıkan gelirlerin ortalamasıyla, B yani yeni sistemin önerilerine göre insanlar ürünleri satın aldığında ortaya çıkan gelirlerin ortalaması birbirine eşittir. 

# Bağımsız İki Örneklem T Testi

- H0: μ1 = μ2
- H1: μ1 ≠ μ2

μ1 eski sistem (A)'yı, μ2 ise yeni sistem (B)'yi temsil etmektedir. Amacımız, bu iki değer arasında bir fark olup olmadığını test etmektir.

## Veri Tipi 1

Genellikle 2 farklı grubun test edilmesi gerektiğinde, elimizde genellikle 'Veri Tipi 1' olarak adlandırdığımız türde veriler bulunur. Bu durumda, elimizde genellikle 'A_B' şeklinde bir veri bulunur.

```python
import pandas as pd
```

```python
# A: Eski sistem
A = pd.DataFrame([30,27,21,27,29,30,20,20,27,32,35,22,24,23,25,27,23,27,23,
        25,21,18,24,26,33,26,27,28,19,25])

# B: Yeni sistem
B = pd.DataFrame([37,39,31,31,34,38,30,36,29,28,38,28,37,37,30,32,31,31,27,
        32,33,33,33,31,32,33,26,32,33,29])
```

```python
A_B=pd.concat([A,B],axis=1) # A ile B'yi yan yana birleştir
A_B.columns=["A","B"] # sütunları adlandırdım.
```

```python
# bu veriyi fonksiyona olduğu gibi yolladığımızda hata alırız.
A_B.head() # bu yüzden bu veriyi veri tipi2'ye dönüştürmeliyiz.
```

```python
import numpy as np
```

```python
# A: Eski sistem
A = pd.DataFrame([30,27,21,27,29,30,20,20,27,32,35,22,24,23,25,27,23,27,23,
        25,21,18,24,26,33,26,27,28,19,25])

# B: Yeni sistem
B = pd.DataFrame([37,39,31,31,34,38,30,36,29,28,38,28,37,37,30,32,31,31,27,
        32,33,33,33,31,32,33,26,32,33,29])
```

```python
# A ve A'nın grubu
GRUP_A = np.arange(len(A)) # 0'dan len(A)'ya kadar (len(A) dahil değil)
GRUP_A = pd.DataFrame(GRUP_A) # DataFrame'e dönüştürdük
GRUP_A[:]="A" # bütün değerleri A yapar. A, A, A...
A=pd.concat([A, GRUP_A],axis=1) # A ile GRUP_A yan yana birleştirdik
A.head()
```

```python
# B ve B'nin Grubu
GRUP_B=np.arange(len(B)) # 0'dan len(B)'ya kadar (len(B) dahil değil)
GRUP_B=pd.DataFrame(GRUP_B) # DataFrame'e dönüştürdük
GRUP_B[:]="B" # bütün değerleri B yapar. B, B, B...
B=pd.concat([B, GRUP_B],axis=1) # B ile GRUP_B yan yana birleştirdik
B.head()
```

```python
# Tüm Veri
AB=pd.concat([A,B]) # A ile B'yi alt alta birleştir.
AB.columns=["gelir","GRUP"] # sütunları isimlendirdim.
```

## Normallik Varsayımı

```python
# Normallik varsayımı için kullanacağımız test: shapiro
from scipy.stats import shapiro
```

```python
# değişkenin normal dağılım varsayımını sağlıyor mu?
# A_B veri setinin içerisinden A değişkenini çekip A'nın normallik testini yaptık
print("A:",shapiro(A_B.A))
# A_B veri setinin içerisinden B değişkenini çekip B'nin normallik testini yaptık
print("B:",shapiro(A_B.B))
```

- shapiro'nun H0 hipotezi istatiksel olarak anlamlı bir şekilde örnek dağılımıyla teorik normal dağılım arasında fark yoktur. 
- A: pvalue < 0.05 olmadığından reddedemeyiz. A için varsayım sağlanmaktadır.
- B: pvalue < 0.05 olmadığından reddedemeyiz. B için varsayım sağlanmaktadır.

## Varyans Homojenliği Varsayımı

- H0: Varyanslar homojendir.
- H1: Varyanslar homojen değildir.

varyans homojenliği için kullanacak olduğumuz test "**levene**"dir.

```python
import scipy.stats as stats
print("Levene test istatistiği:", str(stats.levene(A_B.A, A_B.B)[0]))
print("p-value:", str(stats.levene(A_B.A, A_B.B)[1]))
```

- p-value < 0.05 olmadığından H0'ı reddedemeyiz. Varyanslar homojendir.
- O zaman varyans homojenliği varsayımı sağlanmaktadır.

- Çalışmanın başında kabul ettiğimiz, yani alfa olarak adlandırdığımız, kabul edilebilir hata miktarı 0.05 idi.

## Bağımsız İki Örneklem T Testi: Uygulama

### Hipotez Testi

Bağımsız iki örneklem t testinin hipotezini test edeceğiz.

```python
# A_B isimli veri seti içerisinden A'yı seç, B'yi seç, 
# varyans homojenliği varsayımının sağlanıp sağlanmadığını ifade edelim: True
stats.ttest_ind(A_B["A"], A_B["B"], equal_var=True)
```

- p value < 0.05 olduğundan farklılık vardır. 
- ana kitle ortalamaları birbirine eşittir diyen, farklılık olmadığını ifade eden H0 hipotezi reddedilir.

```python
# .4f: noktadan sonraki 4 değeri göster.
test_istatistigi, pvalue = stats.ttest_ind(A_B["A"], A_B["B"], equal_var=True)
print("Test İstatistiği = %.4f, p-değeri = %.4f" %(test_istatistigi, pvalue))
```

- H0: μ1 = μ2
- H1: μ1 ≠ μ2


- μ1(A): eski sistem, kullanıcıların bir ürün önerisi almadan ürünlerle karşılaşmalarını ifade ediyordu.

- μ2(B): yeni sistem, kullanıcıların geliştirilmiş olan bir ml algoritmasının tahminlerine göre ürünlerle karşılaştırılmasını ifade ediyordu.
- geliştirilmiş olan bir ml modeli aracılığıyla ortaya koyulan tahminlere göre kullanıcıların ürünlerle karşılaştırılmasını ifade ediyordu.


- H0 hipotezi, bu değerlerin birbirine eşit olduğunu ifade ediyordu. p değeri değerlendirdiğimizde H0'ı reddettik. Sonuç olarak, eski sistemle yeni sistem arasında gelirler açısından istatistiksel olarak anlamlı bir farklılık bulunmaktadır. Bu fark, yeni sistemin lehinedir. Çünkü ortalamaları incelediğimizde, yeni sistem olan B'nin, makine öğrenmesine dayalı olarak çalışan grubun daha yüksek bir ortalamaya sahip olduğunu gözlemledik. Bu nedenle, herhangi bir fark varsa, B lehine olacaktır.

# Nonparametrik Bağımsız İki Örneklem T Testi

Varsayalım ki önceki bölümde gerçekleştirmiş olduğumuz iki tane varsayım testinin sonuçları negatif çıktı. Yani hem normallik varsayımı sağlanmıyor, hem de varyans homojenliği varsayımları sağlanmıyor. Bu durumda, bu varsayımların sağlanmadığı durumda kullanılacak olan nonparametrik testimizin adı "mannwhitneyu" olacaktır.

```python
# A_B isimli veri seti içerisinden A'yı seç, B'yi seç
stats.mannwhitneyu(A_B["A"], A_B["B"])
```
