# Poisson Dağılımı

Belirli bir zaman aralığında belirli bir alanda nadiren rastlanan olayların olasılıklarını hesaplamak için kullanılır.

Poisson kesikli bir rassal değişkendir. kesikli bir dağılımdır.

- lambda, çalışmadan bulacak olduğumuz veya daha önce hesaplanmış olan değerler olacak
- E, beklenen değer lambda
- Var, varyans lambda

poisson dağılımında ortalama ve varyans birbirine eşittir. lambda değeridir.

lambda, beklenen sonucun ortalama gerçekleşme sayısıdır. yani ortalamadır. beklenen değer için ortalama ifadesini kullanmıştık yani ortalamadır. beklenen sonucun ortalama gerçekleşme sayısıdır.

- f(x, lambda) buradaki x ise ilgilenmiş olduğumuz olayın ortaya çıkma sayısı

- X rassal değişkeni poisson dağılımına sahiptir. lambda değeri 1'dir.
- beklenen değer ortalamaydı dolayısıyla lambda hem ortalamayı hemde varyansı ifade eder.
- 1, 2, 3 değerlerini aldığında gerçekleşecek olan olasılık değerleri y ekseninde verilmiş.

- Lambdayı yani ortalamayı yani varyansı artırdığımızda yayıldı ve önceki grafikteki soldaki yapının sağa doğru kaymasını bekleriz 

- Lambda daha da arttığında, 10 olduğunda dağılım biraz daha yayıldı. Merkezin etrafında biraz daha farklı değerler gözlenir, çünkü Poisson dağılımının lambda parametresi hem ortalamayı hem de varyansı ifade etmektedir.

1. **Lambda**: Bu terim, Poisson dağılımının parametresidir. Poisson dağılımı, nadir ve rastgele olayların sayısını modeller. Lambda değeri, bu olayların ortalama sayısını ifade eder. Metinde "Lambda daha da arttığında" denilerek, lambda parametresinin değerinin arttığı duruma atıfta bulunuluyor.

2. **Dağılımın Yayılması**: "Dağılımın biraz daha yayıldı" ifadesi, olası sonuçların daha geniş bir aralığa yayıldığı anlamına gelir. Yani daha farklı değerler elde edilme olasılığı artar.

3. **Merkezin Etrafında Değerler**: Dağılımın merkezi, genellikle ortalama veya beklenen değer olarak adlandırılan değere karşılık gelir. Metinde bahsedilen "merkezin etrafında biraz daha farklı değerler gözlenir" ifadesi, daha yüksek lambda değeri nedeniyle olasılık dağılımının merkezinin etrafında daha geniş bir aralıkta değerlerin bulunabileceğini belirtir.

4. **Ortalama ve Varyans**: Lambda parametresi, Poisson dağılımının hem ortalama (beklenen değer) hem de varyans (değişkenlik) değerini ifade eder. Yani lambda değeri arttıkça hem ortalama hem de varyans artar.

Varyans ve ortalamanın birbirine eşit olması, bir olasılık dağılımında değerlerin merkezi konumunun ve dağılımın yayılmasının aynı olduğunu gösterir. Bu durumda, dağılımdaki değerler ortalama etrafında simetrik olarak dağılır ve dağılımın her iki tarafında eşit miktarda değer bulunur.

Örneğin, Poisson dağılımında ortalama ve varyansın birbirine eşit olması, nadir olayların olasılık dağılımında merkezi bir değeri temsil eder. Ancak, genellikle normal dağılım gibi daha yaygın dağılımlarda ortalama ve varyansın tam olarak eşit olması beklenmez.

Bir dağılımda ortalama ve varyansın birbirine eşit olması, dağılımın simetrik olduğunu ve merkezin etrafındaki değerlerin homojen bir şekilde dağıldığını gösterir.

Bir olasılık dağılımında ortalama, dağılımdaki değerlerin merkezi konumunu ifade eder. Varyans ise bu değerlerin ortalama etrafında ne kadar yayıldığını belirler. Varyansın büyük olması, değerlerin ortalama etrafında daha geniş bir aralığa yayıldığını gösterirken, varyansın küçük olması ise değerlerin ortalama etrafında daha yakın bir aralıkta olduğunu gösterir.

Eğer bir dağılımda ortalama ve varyans birbirine eşitse, bu durumda değerlerin ortalama etrafında simetrik bir şekilde dağıldığını ifade eder. Yani, ortalama etrafındaki değerlerin her iki tarafında eşit miktarda değer bulunur. Eşit ortalama ve varyans, dağılımın merkezinin dengeli olduğunu ve her iki yönde benzer miktarda değere sahip olduğunu gösterir.

Bu aynı zamanda dağılımın homojen bir şekilde dağıldığını gösterir. Homojen dağılım, değerlerin merkezden uzaklaştıkça aynı ölçüde yayıldığı ve bu nedenle simetrik bir şekle sahip olduğu anlamına gelir. Varyansın ortalama ile eşit olması, bu homojen dağılımın bir göstergesidir.

### Örnek

- 10 bin kelimeden oluşan bir kitapta hatalı kelime sayısı
- 4000 öğrencili okulda not girişinde hata yapılması
- bir iş gününde çağrı merkezine gelen taktir sayısı
- Kredi kartı işlemlerinde sahtekarlık olması
- Rötara düşen uçuş sefer sayısı

gibi durumlar nadiren gözlenen olaylar olduğundan bu durumda bu dağılım poisson dağılımıdır. poisson ile olasılık hesaplanır. yani gözlem sayısının çok yüksek ve beklenen sonucun gelme olasılığının çok küçük olduğu zamanlarda kullanılır. buna başka bir ifadeyle n: büyük, p: küçük diyebiliriz. n gözlem sayısı, p ise olasılığı ifade eder.

poisson dağılımı, binom dağılımının özel bir halidir.

Bir olayın nadir olay olarak kabul edilebilmesi için genel kabul gören görüş: rassal deneme sayısı (n) n > 50 olmalı ve n * p (olasılık) < 5 olmalıdır (lambda < 5). İki şart aynı anda sağlanmalıdır. eğer bu şartlar sağlanıyorsa bu olaya nadir olay denir.

- Rassal denemeler iki sonuçlu olmalı.
- Aynı koşullar altında gerçekleştirilmelidir.
- Rassal denemeler birbirinden bağımsız olmalıdır.

Rassal denemeler iki sonuçlu olmalı ve hepsi aynı koşullar altında n kez tekrarlanmalıdır. Rassal denemeler birbirinden bağımsız olmalıdır. 
Rassal denemeler bağımsız, iki sonuçlu ve aynı koşullar altında gerçekleşmelidir.

## Poisson Dağılımı

${\displaystyle f(x,\lambda )={\frac {\lambda ^{x}e^{-\lambda }}{x!}}\,\!}$,  x = 0,1,2,...,n

```python
from scipy.stats import poisson
```

```python
lambda_=0.1
```

```python
rv=poisson(mu=lambda_) # poisson nesnesini oluşturduk
```

```python
print("hiç hata olmaması:", rv.pmf(k=0)) # hiç hata olmaması
print("3 hata olması:", rv.pmf(k=3)) # 3 hata olması
print("5 hata olması:", rv.pmf(k=5)) # 5 hata olması
```
