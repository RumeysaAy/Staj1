# Bağımlı İki Örneklem T Testi

Bağımlı iki grup ortalaması arasında karşılaştırma yapılmak istenildiğinde kullanılır.

- Bağımlıya örnek olarak: bir şirkette eğitim verilsin. bu eğitimin öncesinde ve sonrasında bazı gelişmeler ölçülmek istendiğinde yani aynı kitle aynı örneğe iki farklı uygulama yapıldığında ve bunun sonuçları incelendiğinde bağımlılık durumu söz konusudur. bunlara bağımlı gruplar denir.

- Bağımsız ise elimizde 2 tane grup olsun ikiside farklı gelirleri ifade ediyordu ve bu gelirlerin ikiside birbirinden bağımsızdı.

- fakat burada elimizde 1 kitle var. bu kitleye bir işlem uygulanıyor ve sonra tekrar ölçüm yapılıyor. dolayısıyla öncesi ve sonrası birbirine bağımlı

## Bağımlı İki Örneklem T Testi: Hipotezler



önceki durumun(μö) ortalamasıyla, sonraki durumun(μs) ortalaması arasında bir farklılık olup olmadığı ya da yönlü bir farklılık olup olmadığı inceleniyor.

## Bağımlı İki Örneklem T Testi: Test İstatistiği



- buradaki "D"ler farkı ifade eder. öncesi ve sonrası durumlarının birbirinden çıkarılmasını ifade eder. dolayısıyla iki grup olsada bir fark alma işlemi yaparak bu tek gruba düşürülür. tek bir test istatistiğine dönüştürülerek ilgili karşılaştırma işlemleri gerçekleştirilir. 

# Bağımlı İki Örneklem T Testi: Varsayımlar

- Normallik (yaklaşık olarak normal dağılması da kabul görebilmektedir)
- Varyans Homojenliği

# İş Uygulaması: Şirket İçi Eğitimin Performans Etkisi Ölçümü

- Satış departmanı, çalışanlarının satış performansını artırmak amacıyla bir eğitim talep ediyor ve eğitim öncesinde ilgili kişilerin satış performansları ölçülüyor (belirli metriklerle ölçülüyor; bu ortalama satış, birim satış veya günlük ortalama satış gibi belirlenir).
- Sonrasında bir eğitim programı düzenleniyor. Eğitim sonrasında, eğitimin ne kadar işe yarayıp yaramadığını belirlemek amacıyla bir test işlemi gerçekleştirilmek isteniyor.

- **Problem**:
Belirli uğraşlar sonucunda alınan bir eğitimin katma değer sağlayıp sağlamadığı ölçülmek isteniyor.

- **Detaylar**:
    - Bir departman bir konuda eğitim talep ediyor
    - Gerekli/gereksiz değerlendirmeleri neticesinde eğitim alınıyor
    - Eğitimden önce ve sonra olacak şekilde gerekli ölçümler yapılıyor
    - Eğitim sonrasında eğitimin sağladığı katma değer test edilmek isteniyor

# Hipotezimiz



- H0: eğitimin öncesinde katılımcıların ilgili konudaki performans ortalamalarıyla eğitim sonrasında ilgili kişlerin katılımcıların performanslarına ilişkin ortalamalar birbirinden farklı değilidir, birbirine eşittir.

- H1: eğitim etkili olmuştur, bir farklılık oluşmuştur.

# Bağımlı İki Örneklem T Testi

```python
import pandas as pd
import numpy as np
```

```python
# kişilerin eğitimden önceki performans ölçümüne ilişkin değerler
oncesi = pd.DataFrame([123,119,119,116,123,123,121,120,117,118,121,121,123,119,
            121,118,124,121,125,115,115,119,118,121,117,117,120,120,
            121,117,118,117,123,118,124,121,115,118,125,115])

# kişilerin eğitimden sonraki performans ölçümüne ilişkin değerler
sonrasi = pd.DataFrame([118,127,122,132,129,123,129,132,128,130,128,138,140,130,
             134,134,124,140,134,129,129,138,134,124,122,126,133,127,
             130,130,130,132,117,130,125,129,133,120,127,123])
```

- bu veri seti excel, txt ya da csv formunda gelir. bunu python'a okuttuğumuzda DataFrame olur.
- bu veriyle 2 türlü karşılaşabiliriz.

```python
oncesi[0:5]
```

```python
sonrasi[0:5]
```

```python
# BİRİNCİ VERİ SETİ
AYRIK=pd.concat([oncesi, sonrasi], axis=1) # yan yana birleştirir.
AYRIK.columns=["ONCESI","SONRASI"] # sütunları isimlendirdik
print("'AYRIK' Veri Seti: \n\n",AYRIK.head(), "\n\n")
```

```python
# İKİNCİ VERİ SETİ

# ÖNCESİ FLAG/TAG'INI OLUŞTURMA
GRUP_ONCESI=np.arange(len(oncesi))
GRUP_ONCESI=pd.DataFrame(GRUP_ONCESI)
GRUP_ONCESI[:]="ONCESI"

# FLAG VE ÖNCESİ DEĞERLERİNİ BİR ARAYA GETİRME
A=pd.concat([oncesi, GRUP_ONCESI], axis=1) # yan yana birleştirilir.


# SONRASI FLAG/TAG'INI OLUŞTURMA
GRUP_SONRASI=np.arange(len(sonrasi))
GRUP_SONRASI=pd.DataFrame(GRUP_SONRASI)
GRUP_SONRASI[:]="SONRASI"

# FLAG VE SONRASI DEĞERLERİNİ BİR ARAYA GETİRME
B=pd.concat([sonrasi, GRUP_SONRASI], axis=1)


# TÜM VERİYİ BİR ARAYA GETİRME
BIRLIKTE=pd.concat([A,B]) # alt alta birleştirilir. axis=0
BIRLIKTE

# İSİMLENDİRME
BIRLIKTE.columns=["PERFORMANS","ONCESI_SONRASI"]
print("'BIRLIKTE' Veri Seti: \n\n", BIRLIKTE.head(),"\n\n")
print(BIRLIKTE.tail(),"\n\n")
```

```python
# boxplot ile aradaki farklılığı gözlemleyelim
import seaborn as sns
sns.boxplot(x="ONCESI_SONRASI", y="PERFORMANS", data =BIRLIKTE);
```

"SONRASI"nın daha yukarıda konumlandığı, medyanın da daha yukarıda olduğu gözlemlenmektedir. İstatistiksel olarak şu an yorum yapamasak da eğitimin işe yaradığı görülmektedir. Ancak, bunu istatistiksel olarak söyleyemeyiz; şans eseri olmuş olabilir. Bu nedenle, bunu belirlemek için bir test yapmalı ve bundan önce varsayım kontrolü yapmalıyız.

# Bağımlı İki Örneklem T Testi: Varsayım Kontrolü

Bağımlı iki örneklem T testi, bilimsel bir referans noktasını, karşılaştırma noktasını gerekli varsayımlar sağlandığında verir. Çünkü sadece varsayımlar sağlandığında, arka plandaki standartlaştırma işlemleri sonucunda verilebilecek olan kararları bilimsel olarak verebiliriz. Dolayısıyla, bu varsayımları kontrol etmek durumundayız.

## Normallik Varsayımı

```python
# Normallik varsayımını kontrol etmek için shapiro'yu kullanıyorduk.
from scipy.stats import shapiro
```

```python
# Öncesi için normallik varsayımı
shapiro(AYRIK.ONCESI)
```

- Normalliğe ilişkin H0 hipotezi, teorik dağılımla örnek dağılım arasında istatistiksel olarak anlamlı bir farklılık yoktur.
- çalışmanın başında kabul edilebilir hata miktarı olarak kabul ettiğimiz alfa, 0.05
- pvalue < 0.05 sağlanmadığından H0 hipotezini reddedemeyiz bu şu anlama gelir demek ki elimizdeki örneğin dağılımı normaldir

```python
# Öncesi için normallik varsayımı
shapiro(AYRIK.SONRASI)
```

- pvalue < 0.05 sağlanmadığından H0 hipotezini reddedemeyiz.
- bu teorik normal dağılım ile örnek dağılım arasında bir fark yoktur. yani normal dağılmaktadır.
- Normallik varsayımı, iki grup içinde sağlanmaktadır.

## Varyans Homojenliği

- H0: Varyanslar homojendir.
- H1: Varyanslar homojen değildir.

```python
import scipy.stats as stats
stats.levene(AYRIK.ONCESI, AYRIK.SONRASI)
```

- p değeri 0.05'ten küçük olduğunda, H0 hipotezini reddederiz. Varyans homojenliği varsayımı sağlanmamaktadır.


Bu durumda yapılabilecekler:
- veri seti üzerinde bazı aykırılıklar varsa bunlarla ilgili düzenleme işlemleri yapılarak bu varsayımlar tekrar incelenebilir.
- bağımlı iki örneklem t testinde varyans homojenliği incelendiğinde burada bu varsayımın sağlanmaması durumunda bu bir miktar göz ardı edilebilen bir durumdur.

Varyans homojenliği varsayımı sağlanmışcasına ilgili testimizi gerçekleştireceğiz.

# Bağımlı İki Örneklem T Testi: Uygulama
## Hipotez Testi

```python
stats.ttest_rel(AYRIK.ONCESI, AYRIK.SONRASI)
```

```python
test_istatistigi, pvalue =stats.ttest_rel(AYRIK["ONCESI"], AYRIK["SONRASI"])
print("Test İstatistiği = %.5f, p-değeri = %.5f" % (test_istatistigi, pvalue))
```

- H0: çalışanlarımıza eğitim vermeden önceki performanslarıyla eğitim aldıktan sonraki performansları karşılaştırıldığında bu performansların ortalamaları arasında istatiksel olarak anlamlı bir farklılık yoktur.
- p-value < 0.05 sağlandığından H0 hipotezi reddedilir. yani eğitim işe yaramıştır. 

çalışanların satış performansları göz önünde bulundurulduğunda öncesi ve sonrası ölçümleri performansları incelendiğinde istatistiksel olarak anlamlı bir farklılık, %5 hata payı ve %95 güvenirlilik ile ortaya konmuştur. eğitim başarılıdır.  

# Nonparametrik Bağımlı İki Örneklem T Testi

Varsayalım ki, varsayımlar bölümündeki normallik varsayımı ve varyans homojenliği varsayımı sağlanmadı.

Bu durumda wilcoxon testini kullanacağız. wilcoxon testi nonparametrik bağımlı iki örneklem testidir.
stats'ın içerisinde yer almaktadır.

```python
stats.wilcoxon(AYRIK.ONCESI, AYRIK.SONRASI)
```

```python
test_istatistigi, pvalue = stats.wilcoxon(AYRIK["ONCESI"], AYRIK["SONRASI"])
print("Test İstatistiği = %.4f, p-değeri = %.4f" % (test_istatistigi, pvalue))
```

- p-value < 0.05 olduğundan H0 reddedilir. iki grup arasında istatistiksel olarak oldukça anlamlı bir farklılık gözlemlenmektedir. verilen eğitim performansları çok olumlu bir şekilde etkilemiştir.

- 2'li grup çalışmalarında yani bağımsız iki örneklem t testinde, bağımlı iki örneklem t testinde  elimizde aslına bakarsanız 3 tane hipotez türümüz var. eşit değildir, büyüktür ya da küçüktür şeklinde alternatif hipotezlerimiz vardı.

- tek örneklem testlerinde belirli bir skaler değerden büyüktür ya da küçüktür şeklinde alternatif hipotezlerimiz vardı.

- bu alternatif hipotezlerde python dünyasında sık kullanılan kütüphanelerin içerisindeki kullanılan fonksiyonların yön konusunda ya desteği yok ya da bazı problemleri var. bundan dolayı çok iyi çalıştığını bildiğimiz direkt alternatif hipotezleri eşit değildir şeklinde yapıyoruz. bu makuldür. çünkü zaten ortalamasına baktığımız iki grubu karşılaştırdığımızda hangi grubun yukarıda olduğunu biliyoruz. dolayısıyla iki yönlü farklılık testi yaptığımızda yani alternatif hipotezimizi H1 hipotezimize "öncesi farklıdır sonrası" bunu test ettiğimizde eğer H0 hipotezini reddediyorsak ortalamalar farklıdır anlamına gelir. bu farklılığın hangi grubun lehine olacağı da açıktır. boxplot'da baktığımızda yüksekteki grup "sonrası"ydı. farkı ispat ettiysek zaten farklılık "sonrası" grubundan kaynaklıdır.

- yani kolaylık olması açısından yönlerle hiç ilgilenmeyeceğiz. hipotezlerimizin hepsini iki yönlüymüşcesine düşünüp bu şekilde testini yapacağız. test sonucunda yorumlarımızı yüksek olduğunu bildiğimizin lehine olacak şekilde gerçekleştiriyoruz.
