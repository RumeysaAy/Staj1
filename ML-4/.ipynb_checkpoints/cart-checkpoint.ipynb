# CART - Model & Tahmin

```python
import pandas as pd
# veri setimizi import edelim
diabetes = pd.read_csv("diabetes.csv")
```

```python
df = diabetes.copy()
df = df.dropna()
y = df["Outcome"]
X = df.drop(["Outcome"], axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.30,
                                                   random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# modeli kuralm
cart = DecisionTreeClassifier()
cart_model = cart.fit(X_train, y_train)
```

```python
from skompiler import skompile
# karar kuraln karalm
# tahmin fonksiyonunu karsn ve python kodlarna dntrsn
print(skompile(cart_model.predict).to("python/code"))
```

```python
# tahmini y test deerleri
y_pred = cart_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
# doruluk oran
accuracy_score(y_test, y_pred)
```

# CART - Model Tuning

```python
cart_model
```

```python
?cart_model
```

* min_samples_split: blnme iin gerekli olan min rnek says
* min_samples_leaf: leaf node dediimiz en son node'larda barndrlacak olan eleman saysn ifade eder. n tanml deeri 1'dir

```python
cart_grid = {"max_depth": list(range(1,10)),
            "min_samples_split": list(range(2,50))}
```

bu iki argman dallanma kontrolu salar. ar renmeyi engeller ve rassall korur.

```python
from sklearn import tree
cart = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
cart_cv = GridSearchCV(cart, cart_grid, cv=10, n_jobs=-1, verbose=2)
```

```python
cart_cv_model = cart_cv.fit(X_train, y_train)
```

```python
print("En iyi parametreler: " + str(cart_cv_model.best_params_))
```

```python
# bu parametreleri kullanarak final modelimizi oluturalm
cart = tree.DecisionTreeClassifier(max_depth =5, min_samples_split =19)
# min_samples_split: blnme iin gereken rnek/gzlem says
```

```python
cart_tuned = cart.fit(X_train, y_train)
```

```python
# y test tahmini deerleri
y_pred =cart_tuned.predict(X_test)
```

```python
# doruluk oran
accuracy_score(y_test, y_pred)
```
