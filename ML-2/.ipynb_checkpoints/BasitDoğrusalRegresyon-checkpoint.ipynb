# statsmodels kütüphanesiyle ilk modellememizi yapalım.
## basit doğrusal regresyon modeli
### tek değişkenli bir modelleme yapacağız

```python
import statsmodels.api as sm
```

```python
X = df[["TV"]] # X, bağımlı değişkendir
X[0:5]
```

elimizde tek değişkenden oluşan bir matris var. TV değişkenini X matrisi olarak kaydettik. burada bu regresyon probleminin çözülmesi için elimizdeki bu X matrisine tek değişken olmasından dolayı ve matris işlemlerinin yapılabilmesi için değişken olarak "1" ifadesini ekleyeceğiz.

```python
X = sm.add_constant(X) # 1 değerlerinden oluşan bir sabit değişken ekledim
X[0:5]
```

X bağımsız değişkenimiz hazır

```python
# bağımlı değişkenimizi alalım
y=df["sales"]
y[0:5]
```

```python
# modeli kuralım.
# lineer model
# y bağımlı değişken, X bağımsız değişken
lm = sm.OLS(y,X)
```

- AIC ve BIC gibi model karşılaştırma için kullanılan iki istatistik değeri var.
- Df Residuals:	198 (serbestlik derecesi)

```python
model=lm.fit()
```

```python
model.summary()
```

modelde kullandığımız bağımsız değişkenimizin(TV) katsayısı ve sabit(const) katsayı ifadelerinin yorumlanmalarına geçelim
- const & coef: 7.0326 (beta0 değeri)(küçük)
- coef: katsayı, parametre
- parametre konusunu regresyon problemleri anlamında ele alırsak parametre, teorik olarak gösterdiğimiz anakitleye ilişkin beta0 ile beta1 değerleri(büyük)
- katsayılar ise model üzerinden hesaplamış olduğumuz bu tablodaki coef(katsayılar)(küçük beta0 ve küçük beta1)
- std err: katsayının standart hatası verilmiş
- bunların anlamlı olup olmadığını anlamak için t-istatistiği(t) ve p-value>t verilmiş 
- p-value>t baktığımızda TV ve const'un coef(katsayıları) anlamlı olduğu gözükür.
- [0.025	0.975]: bu katsayıların güven aralığıdır.
- bu test çıktısı hesaplamış olduğumuz TV'ye ilişkin 0.04 şeklinde olan katsayımız istatistiksel olarak anlamlıdır ve yine bu katsayı %95 güvenirlilik ile güven aralığında(0.042 0.053) yer alacaktır.
- sabit(const) eğrinin y eksenini kestiği nokta coef(7.0326). diğer değişkenlerde hiçbir değişiklik olmadığında y'nin beklenen değeri, ortalama satıştır. yani TV'nin bir etkisi olmadığında satışlarda zaten olmasını beklediğimiz ortalama satış miktarı 7.0326'dır.
- TV harcamalarında bir birimlik artış olduğunda, satışlarda ortalama 0.0475 kadar bir artış olması beklenir.
- her bir değişkenin bağımlı değişkene olan etkilerini görüyoruz(coef)

- Dep. Variable: sales (bağımlı değişken)
- No. Observations:	200	(gözlem sayısı)

- R-squared: 0.612 (R-kare değeri) bağımsız değişkenin, bağımlı değişkendeki değişkenliği açıklama başarısıdır. yani burada elimizdeki TV değişkeni, satış değişkeninde yer alan değişkenliğin yaklaşık %60'ını açıklayabilmektedir.
- Adj. R-squared: 0.610 (düzeltilmiş R-kare değeri) ile R-squared arasındaki fark R-squared değeri formülasyonu itibariyle modele ne kadar fazla değişken eklersek ekleyelim şişmeye meyillidir, artacaktır. dolayısıyla modelde ilgili ilgisiz, anlamlı anlamsız değişken olması durumunda R-squared artacaktır. R-squared açıklanabilirlik oranıdır. dolayısıyla modelde ilgili ilgisiz birçok değişken ekleyip R-squared'nin arttığını gördüğümüzde iyi bir yolda olduğumuzu düşünebiliriz bu çok yanıltıcı bir durumdur. bu düzeltilmiş R-squared, R-squared'nin her parametre eklenmesine karşı olan duyarlılığını törpüleyen düzenleyen duyarlılığı daha az olan bir metriktir. dolayısıyla burada şu an bu problem için çok geçerli olmasa da birden ve ikiden fazla belki onlarca değişken olduğunda düzeltilmiş R-squared değeri bizim için daha düzgün bir değerdir. 

- F-statistic: 312.1 modelin anlamlılığının anlaşılması için kurulan test istatistiğidir.
- Prob (F-statistic): p-value değerinin kendisidir.
- F-statistic ve Prob incelendiğinde modelin anlamlı olduğu gözlemlenmektedir.

# Artıklar ve Makine Öğrenmesindeki Önemi

```python
# hata kareler ortalaması, R-squared
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# import statsmodels.formula.api as smf
# statsmodels üzerinden model kurduk
lm = smf.ols("sales ~ TV", df)
model = lm.fit()
```

```python
# hata kareler ortalaması
# gerçek değerler ile tahmin edilen değerler arasındaki farkların karelerinin ortalaması
# gerçek değerler, tahmin edilen değerler
# y, bağımlı değişken
mse=mean_squared_error(y, model.fittedvalues)
mse # birim başına yapacak olduğumuz ortalama hata
```

```python
# hata kareler ortalamasının karekökü
import numpy as np
rmse = np.sqrt(mse)
rmse 
```

```python
# sklearn üzerinden bu modeli kurmuştuk LinearRegression
# tahmin edilen değerler
# reg nesnesi üzerinden tahmin yap
reg.predict(X)[0:10]
```

```python
# gerçek değerler
y[0:10]
```

```python
# karşılaştırma tablosu
import pandas as pd
k_t = pd.DataFrame({"gercek_y":y[0:10],
                   "tahmin_y": reg.predict(X)[0:10]})
k_t
```

```python
# her bir gerçek y değerine karşılık tahmin etme işlemi gerçekleştirdikten sonra
# yapacağım hata nedir?
k_t["hata"] = k_t["gercek_y"] - k_t["tahmin_y"]
k_t
```

```python
# birim başıma ne hata yaptığım bilgisine erişmek istiyorum
# ortalama ne kadar hata yaptım ?
# eksi değerler artıları götüreceğinden bu yüzden karesini alalım. (mutlak değerde alınabilir)
k_t["hata_kare"] = k_t["hata"]**2
k_t
```

```python
# toplam hata
np.sum(k_t["hata_kare"])
```

```python
# hata kareler ortalama
np.mean(k_t["hata_kare"])
```

```python
# hata kareler ortalama karekökünü alalım
np.sqrt(np.mean(k_t["hata_kare"]))
```

Gerçek değerler ile tahmin değerleri arasındaki hataları bir şekilde her zaman optimize etmeye çalışmak olacaktır. amacımız hatayı min yapacak şekilde ilgili kural setlerine katsayılara ulaşmak olacaktır.

artıkların değerlendirilmesiyle ilgili kritik bilgi: bir model projesi gerçekleştirdiniz ve tahmin değerlerini, başarı değerlerini örneğin mse veya rmse değeri 9 çıktı. hatayı bu şekilde muhattaplara sunmak yanıltıcıdır, aldatıcıdır. hata şu şekilde sunulmalıdır: hataların dağılımına bakılmalıdır. örneğin hata değeri 9 ama hataların grafiğini ortaya döktüğümüzde dağılımına baktığınızda veri setinin gözlemleri söz konusu olduğunda bu veri setinin %95'inde 0.01 hata yaptım geri kalan %5inde şu kadar hata yaptım
- hatanın tanımlanma şekli değişti rmse, mse değil mse, rmse kıyaslama için kullandığımız temel değerler. bu temel değerlere bağlı olarak bir işin başarılı ya da başarısız olduğunu değerlendirmek gerçek hayatta çok doğru olmayacaktır. lehimize veya aleyhimize olabilir
- dolayısıyla elde ettiğimiz hataları görselleştirmek başarılı ve başarısız olduğunuz kısımları incelememiz gerekir.
- örneğin bahsettiğimiz %95'lik kısımda 0.01 hatayla çalıştınız başarılısınız
- geriye kalan %5lik kısıma odaklanıp bu gözlemlerin kendilerine gidip yapısal bir problem olup olmadığını incelememiz gerekir
- bunların taşıdığı karakteristiği? neden böyle aykırılıklar mı yoksa başka bir problem mi bunların neden başarılı bir şekilde tahmin edemediğinize gidip odaklanmanız gerekir.

```python
# modelin artıkları(hataları)
model.resid[0:10]
```
