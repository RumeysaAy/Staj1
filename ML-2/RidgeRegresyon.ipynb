# Ridge Regresyon - Model

```python
import pandas as pd
import numpy as np
# from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.model_selection import train_test_split
```

```python
hit = pd.read_csv("Hitters.csv")
```

```python
df = hit.copy()
```

```python
df = df.dropna()
```

```python
dms = pd.get_dummies(df[["League","Division","NewLeague"]])
```

```python
y = df["Salary"]
```

```python
X_ = df.drop(["Salary","League","Division","NewLeague"], axis=1).astype("float64")
```

```python
X = pd.concat([X_, dms[["League_N","Division_W","NewLeague_N"]]], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import Ridge
```

```python
# lambda = 0.1
ridge_model = Ridge(alpha=0.1).fit(X_train, y_train)
ridge_model
```

```python
# modelin katsayıları
ridge_model.coef_
```

```python
# farklı lambda değerlerine göre bu katsayıları gözlemleme işlemini gerçekleştirelim.
# döngü yazalım

# lambdalar isimli bir array oluşturduk
# başka değerleri ilede oluşturulabilir
lambdalar = 10**np.linspace(10,-2,100)*0.5
# lambdalara göre katsayıların nasıl değiştiğini gözlemlemeye çalışacağım
# hataları değil katsayıları gözlemleyeceğim
# Ridge ve Lasso Regresyon modeli öncekilerden farklı olarak katsayılarla uğraşıyoruz
# katsayılara ceza uygulama işlemi yapıyoruz bu ceza işleminin şiddetini etkileyen şey 
# bizim kontrolumuzde olan lambda değeri dolayısıyla bu lambda değerine karşılık 
# bu katsayılarda nasıl değişiklikler meydana geldiğini bir gözlemleyelim 
# lambda belirlemenin önemini anlayalım.

ridge_model = Ridge() # model nesnesi oluşturdum
katsayilar=[]

for i in lambdalar:
    # her iterasyonda verilen lambda değerlerini parametre olarak alacak
    ridge_model.set_params(alpha=i)
    ridge_model.fit(X_train, y_train)
    # katsayilar array'ine her iterasyonda katsayıları ekledi
    # her lambda'nın 19 tane katsayısı olacak (19 tane değişken var)
    katsayilar.append(ridge_model.coef_)
```

* lambdalar = 10**np.linspace(10,-2,100)*0.5

İfade, bir dizi sayının oluşturulmasını içerir ve bu sayıların hesaplanma mantığını temsil eder. İşte ifadenin ayrıntılı açıklaması:

1. `np`: Bu muhtemelen "NumPy" kütüphanesini temsil eder. NumPy, Python programcılarının bilimsel hesaplamalar yapmak için kullanabileceği bir kütüphanedir. NumPy, özellikle çok boyutlu diziler ve matrisler üzerinde işlem yapmak için kullanılır.

2. `np.linspace(10, -2, 100)`: Bu NumPy fonksiyonu, belirtilen aralıkta belirli bir sayıda noktayı içeren bir dizi oluşturur. Burada, aralık 10'dan -2'ye kadar olan sayıları içerir ve toplamda 100 nokta oluşturulur. Yani, 10 ile -2 arasında eşit aralıklı 100 adet sayı içeren bir dizi oluşturulur.

3. `10**np.linspace(10, -2, 100)`: Bu ifade, önceki adımda oluşturulan her bir sayının 10 üzeri alınmasını temsil eder. Yani, dizideki her bir sayı 10'un üssü olarak hesaplanır.

4. `10**np.linspace(10, -2, 100) * 0.5`: Bu ifade, önceki adımda hesaplanan her bir sayının 0.5 ile çarpılmasını ifade eder. Yani, dizideki her bir sayı 0.5 ile çarpılır.

Sonuç olarak, bu kod parçası `lambdalar` adında bir dizi oluşturur. Bu dizi, 10'un -2'ye kadar olan aralıkta eşit aralıklı üslerini (10 üzeri) alarak oluşturulur ve her bir üs değeri 0.5 ile çarpılır. Bu dizinin elemanları, lambda (λ) adı verilen bir parametrenin çeşitli değerlerini içerir. Genellikle matematiksel veya bilimsel hesaplamalarda kullanılan bu tür ifadeler, belirli aralıklardaki parametre değerlerini oluşturmak için kullanılır.

5. `plt.title("Düzenlileştirmenin Bir Fonksiyonu Olarak Ridge Katsayıları")`: Bu satır, grafiğin başlığını belirler. Başlık, "Düzenlileştirmenin Bir Fonksiyonu Olarak Ridge Katsayıları" olarak ayarlanır.

Sonuç olarak, bu kod parçası Ridge regresyonunun düzenlileştirme etkisini ve lambda değerlerinin değişimiyle ağırlıkların nasıl değiştiğini görselleştirmek için kullanılır. Logaritmik x ekseni ölçeği, lambda değerlerinin geniş bir aralığını daha iyi görselleştirmek için kullanılır. Etiketler ve başlık, grafiğin daha iyi anlaşılmasına yardımcı olur.

```python
# grafik ile ilgili ayarlamalar
ax = plt.gca()
ax.plot(lambdalar, katsayilar)
ax.set_xscale("log") # x eksenini ölçeklendirme çıkan değerleri 
# logaritmik dönüşüme sokmalıyız
# görselin yorumlanması, değerlendirilmesinin daha iyi olması için

plt.xlabel("Lambda(Alpha) Değerleri")
plt.ylabel("Katsayılar/Ağırlıklar")
plt.title("Düzenlileştirmenin Bir Fonksiyonu Olarak Ridge Katsayıları");
```

lambda değerlerinin vermiş olduğumuz aralıktaki değişmelerine göre katsayıların yani 
ağırlıkların buradaki değişimleri gözlemlenmiş.

lambdayı ridge ve regresyon anlamında kavranması : 
lambda değerleri değiştiğinde bu renklerin her birisi ayrı bir parametreyi temsil ediyor. lambda değerleri verildikten sonra yavaş yavaş katsayıları 0'a yaklaşıyor.

Ridge regresyonun lambda muamelesi sayıları hiçbir şekilde 0 yapmamak ama bütün katsayıları modelde tutup bu katsayıların modele olan etkilerini ayarlayabilmek

0'a yaklaştırmak mesela (Ridge ile Lasso regresyonun farklılaştığı nokta)

dolayısıyla elimizde yine bildiğimiz bir regresyon katsayılar olayı var farkı bu katsayılara ceza uygulanması ama bu katsayılara belirlenecek lambda değerine göre ceza uyguluyoruz. bu katsayılar bazen etkilerini önemlerine göre kaybediyor. seçilecek olan lambdaya göre değişkenlerin görece etkilerini tahmin fonksiyonundaki etkilerini biz ayarlayabilmiş oluyoruz ama modelde kalsın sıfıra yaklaştıralım modelden çıkarmayalım ridge modelde

# Ridge Regresyon - Tahmin

### Test hatası

```python
y_pred = ridge_model.predict(X_test)
y_pred[:10] # y tahmin değerleri
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# rmse
# lambda = 0.1 e göre test hatası hesapladık
np.sqrt(mean_squared_error(y_test, y_pred))
```

# Ridge Regresyon - Model Tuning

Amacımız optimum lambda değerlerine ulaşmak olacaktır.

veri görselleştirme yaparak katsayıları gözlemledik. çeşitli lambda değerlerine karşılık katsayı değerlerinin neler olacağının gözlemlenmesiydi.

şimdi bu lambda değerlerinden hangisinin daha iyi olduğunu model tuning bölümünde seçmiş olacağız

```python
lambdalar = 10**np.linspace(10, -2, 100)*0.5
lambdalar[0:5]
```

```python
from sklearn.linear_model import RidgeCV
```

```python
# Cross Validation yöntemiyle bu seçme işlemini gerçekleştireceğiz
# valide edilmiş model
ridge_cv = RidgeCV(alphas = lambdalar, 
                   scoring = "neg_mean_squared_error", 
                   normalize = True) # normalleştirme yaptım
```

```python
ridge_cv.fit(X_train, y_train)
```

```python
# optimum lambda değeri
ridge_cv.alpha_
```

lambda'yı çağırdığımda iteratif işlemleri, döngü yapmadık (brute force)

burada ise sklearn'nün içerisinde yer alan ridge regresyonda CV yöntemiyle optimum lambda değerini vermek üzere hazırlanmış olan fonksiyonu argümanlarını tanımlayarak kullandık ve optimum lambda değerini elde ettik

elde ettiğimiz bu optimum lambda değeriyle final modelini oluşturacağız

```python
# tun edilmiş modeli oluşturduk
ridge_tuned = Ridge(alpha = ridge_cv.alpha_, # optimum lambda değeri
                   normalize = True).fit(X_train, y_train)
```

```python
# tahmin işlemini gerçekleştirelim
```

```python
# tahmin edilen y değerleri
y_predict = ridge_tuned.predict(X_test)
```

```python
# rmse test hatası
np.sqrt(mean_squared_error(y_test, y_predict))
```
