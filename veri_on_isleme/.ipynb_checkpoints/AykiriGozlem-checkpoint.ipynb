# Aykırı Gözlem

Veride genel eğilimin oldukça dışına çıkan ya da diğer gözlemlerden oldukça farklı olan gözlemlere _**aykırı gözlem**_ denir.

Aykırılığı ifade eden nümerik değere _**aykırı değer**_ denir.

Aykırı değeri barındıran gözlem birimine _**aykırı gözlem**_ denir.

Aykırı değerlere outlier, aşırı değer, aşırı gözlem, uç değer, sıra dışı değer, aykırı değer denilebilir.

Aykırı gözlem, uğraşacak olduğumuz problemlerin içerisinde en ciddi problemlerden birisidir.

Veri ön işleme bölümünde veriyi ön işleme işlemlerinden geçireceğiz. Bu işlemlerin arasında aykırı gözlen çok özel bir yere sahiptir.

## Aykırı Değer Neye Sebep Olur?

_**Genellenebilirlik**_ kaygısı ile oluşturulan **kural setlerini** ya da **fonksiyonları** yanıltır. **Yanlılığa** sebep olur.



- sol taraftaki görselde hiçbir aykırı gözlem, aykırı değer bulunmamaktadır. Bu durumda x ve y ekseni incelendiğinde x ve y arasındaki ilişkiyi ifade eden mavi doğru eğimi negatif
- sağ taraftaki grafikte ise mavi doğrunun yönü değişmiştir. çünkü grafiğin üst tarafındaki 3 nokta doğrunun yönünü değiştirmiştir.
- değişkenler arasındaki ilişkiyi ifade eden doğrunun aykırı gözlemli senaryoda yönü değişmiştir. Bu sebeple aykırı gözlemler, kendisinden kaçmaya çalıştığımız veya kendisini tamir etmeye, düzeltmeye çalıştığımız çeşitli yöntemler kullandığımız ve bizim için çok önemli bir yere sahip olan bir problemdir.
- soldaki örnekte x ve y arasındaki ilişkiyi tanımlamak için negatif yönlü, negatif eğime sahip bir mavi doğru kullanılmışken, sağ taraftaki aykırı gözlemler yüzünden verinin geneliyle ilgili yapıyı temsil etmek için oluşturulan mavi doğrunun eğimi, yönü değişmiştir. işte aykırı gözlemler bu derecede ciddi şekilde bazı yanlılıklara sebep olabilir.

## Neye Göre Kime Göre Aykırı Gözlem?

**Veride genel eğilimin oldukça dışına çıkan gözlemler**

Peki veri setinin genel eğiliminin dışına çıkmayı nasıl tanımlarız? Aykırı gözlem nasıl tanımlanır?

Bazı senaryolarda aykırı gözlemleri nümerik yöntemlerle tanımlayamayız, çalışıyor olduğumuz sektörün dinamiklere göre neyin aykırı ya da neyin normal olduğu durumunu tanımlayabiliriz. 

**1. Sektör Bilgisi**
Örneğin bir ev fiyat tahmin modelinde, 1000 metrekarelik evleri modellemeye almamak.

bir sektörde bir kişi neyin aykırı olduğunu tanımlayabilir.

makine öğrenmesi çalışmalarında amacımız genellenebilir veri setinin içerisindeki yapıları temsil etme kabiliyeti yüksek yansız modeller oluşturmaktır.

1000 metrekare ve üzeri evler az bulunduğundan bu evleri bir tahmin modeli için çalışmaya dahil etmek diğer daha düşük metrekarelerde olan evler için yapılacak olan genellemelere zarar verecektir. çünkü kullanıyor olduğumuz fonksiyonel yapılar oradaki 1000 metrekarelik evlerle ilişkili yapılarla alakalıda bazı genellenebilirlikler sağlamaya çalıştığından dolayı bu durumda veri setinin içerisindeki diğer yapıları kaçırabilmektedir.

Eğer kurulan modelin bir genelleme kaygısı varsa; zaten çok seyrek olan senoryalar ve genele uymayan yapılar çalışmanın dışında bırakılmalıdır.

çünkü kullanacak olduğumuz fonksiyonlar genelleme kaygısıyla örneğin buradaki örnekteki 1000 metrekarelik evlere ilişkin bilgiyide modellemeye çalıştığından verinin genelindeki bazı yapıları iyi temsil edemez. bu da problemlere sebep olur.

**2. Standart Sapma Yaklaşımı**
Bir değişkenin ortalamasının üzerine aynı değişkenin standat sapması hesaplanarak eklenir. 1, 2 ya da 3 standart sapma değeri ortalama üzerine eklenerek ortaya çıkan bu değer eşik değer olarak düşünülür ve bu değerden yukarıda ya da aşağıda olan değerler aykırı değer olarak tanımlanır.

Araç Fiyatı: 10000, 54000, 46999, 89000, 70000, 50000, NA, 68900, 12000

- Ortalama = 100,000
- Standart Sapma = 20,000

- Eşik Değer = ortalama + 1*(Standart Sapma)
- Eşik Değer = ortalama + 2*(Standart Sapma)
- Eşik Değer = ortalama + 3*(Standart Sapma)

Eşik Değer = 100,000 + 2*20,000 = 140,000

ortalamanın üzerine standart sapma eklendi. yani merkezi eğilim var, o eğilimin etrafındaki dağılımı da göz önünde bulundurduk dolayısıyla merkezi eğilimin üzerine dağılımıda koyup merkezi eğilim artı dağılımın üzerinde kalanlarla ilgilenmiyoruz. çünkü bunlar veri setinin içerisindeki değişkenliğin dışında kalan değerlerdir.

**3. Z-Skoru Yaklaşımı**
Standart sapma yöntemine benzer şekilde çalışır. Değişken standart normal dağılıma uyarlanır, yani standartlaştırılır. Sonrasında -örneğin- dağılımın sağından ve solundan -/+ 2.5 değerine göre bir eşik değer konulur ve bu değerin üzerinde ya da altında olan değerler aykırı değer olarak işaretlenir.



- standartlaştırma işlemi gerçekleştirilir.
- sonrasında dağılımın sağından veya solundan +/- 2.5 değerine göre bir eşik değer konulur. bu değerin üzerinde veya altında olan değerler aykırı değer olarak işaretlenir.
- bu yaklaşımda aslında değişkenin genel yapısının üzerine, değişkenin genel yapısının dağılımını da ifade eden +/- bir değer koyarak veri setinin genel dağılımının dışındaki yapılara erişmeye çalışılır.
- standartlaştırılmış dağılımın sağından ve solundan -/+ 2.5 değerleri

**4. Boxplot (interquartile range - IQR) Yöntemi**
En sık kullanılan yöntemlerden birisidir. Değişkenin değerleri küçükten büyüğe sıralanır. Çeyrekliklerine(yüzdeliklerine) yani Q1, Q3 değerlerine karşılık değerler üzerinden bir eşik değer hesaplanır ve bu eşik değere göre aykırı değer tanımı yapılır.

boxplot grafiği min değer, max değer, 1 ve 3. çeyrekteki değer ve medyana göre oluşturuluyor. burada bu yöntem kullanılarak veri setinin genel yapısının dışında kalan değerlere erişilir. burada eşik değerleri hesaplamak için
- IQR = 1.5*(Q3 - Q1)
- Alt Eşik Değer = Q1 - IQR
- Üst Eşit Değer = Q3 + IQR

belirlenmiş olan bu sınırların üzerinde veya aşağısında kalan değerlere aykırı değer muamelesi yapılır. 

elimize bir veri seti geldiğinde bu veri setinin değişkenlerini olduğu şekliyle bırakıp hemen bir modelleme yapmak yerine veri setinin içerisindeki değişkenlerin aykırılık durumunu eksik gözlem durumunu ve benzeri durumlarını incelememiz gerekir.

eğer aykırı gözlem göz önünde bulundurulmazsa bir doğrunun yönünü değiştirebilir. bu sebeple odaklanılması gerekir.
