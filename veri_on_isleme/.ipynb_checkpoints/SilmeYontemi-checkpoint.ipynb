# Silme Yöntemleri

```python
import numpy as np
import pandas as pd
```

```python
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
```

```python
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
```

```python
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
```

```python
df=pd.DataFrame({
    "V1":V1,
    "V2":V2,
    "V3":V3
})

df
```

```python
# en az bir eksik değere sahip gözlemin silinmesi
df.dropna()
```

```python
df
```

```python
# sadece bütün değerleri NaN olan gözlem birimlerini silmek için
# hepsi aynı anda eksikse sil
df.dropna(how="all")
```

```python
df
```

```python
df["V5"]=3.3
df
```

```python
# değişkenler bazında silme işlemi yapalım.
# en az bir tane eksik değere sahip olan değişkeni sil
df.dropna(axis=1)
# tüm değerleri dolu olan değişken gösterilir.
```

```python
# "sil_beni" adında bir değişken oluşturalım. 
# bu değişkenin tüm değerlerine NaN atayalım.
df["sil_beni"]=np.nan
df
```

```python
# değişkenlere göre işlem yap: axis=1
# tüm değerleri nan olan değişkenleri siler: all
# kalıcı olması için: inplace=True
# tüm değerleri NaN olan değişkenleri, sütunları silelim.
df.dropna(axis=1, how="all",inplace=True)
```

```python
df
```

```python
df.drop("V5", axis=1, inplace=True)
```

```python
df
```
